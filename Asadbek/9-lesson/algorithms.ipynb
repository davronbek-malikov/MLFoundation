# get to know data

```python
import pandas as pd

df=pd.read_csv("covid_ml.csv")
```

```python
df.info()
```

```python
df.head(10)
```

```python
df.drop("ID", axis=1, inplace=True)
```

```python
df.info()
```

# Missing values


```python
df.isnull().sum()
```

```python
for col in df.columns:
	if df[col].isnull().any():
		if df[col].dtype == "object":
			df[col].fillna(df[col].mode()[0], inplace=True)
		else: df[col].fillna(df[col].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df.info()
```

# Encoding

```python
df.nunique()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()

for col in df.columns:
	if df[col].dtype == "object":
		df[col]=encoder.fit_transform(df[col])
		# if df[col].nunique() <= 5 and col != "urgency_of_admission":
		# 	dummies=pd.get_dummies(df[col], prefix="col", dtype=int)
		# 	df=pd.concat([df.drop(columns=[col]), dummies], axis=1)
		# else: df[col]=encoder.fit_transform(df[col])
```

```python
df.info()
```

```python
df.head(10)
```

# Scaling

```python
from sklearn.preprocessing import MinMaxScaler

scaler=MinMaxScaler()

num_col=df.select_dtypes(include=['int64','float64']).columns.drop("urgency_of_admission")
for col in num_col:
    df[col]=scaler.fit_transform(df[[col]])
    
```

```python
df.head(10)
```

# X, Y (inputs and output)

```python
x=df.drop("urgency_of_admission", axis=1)
# y=df["urgency_of_admission"].astype(int)
y=df["urgency_of_admission"]
```

```python
x
```

```python
y
```

# Splitting (train and test)

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x, y, test_size=0.2, random_state=42)
 
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

# Trainig(fitting)

```python
from sklearn.linear_model import LogisticRegression

log_reg=LogisticRegression()
```

```python
log_reg
```

```python
log_reg.fit(x_train, y_train)
```

# Predicting

```python
y_predict=log_reg.predict(x_test)
```

```python
y_predict
```

```python
y_predict[10:20]
```

# Evaluation

```python
from sklearn.metrics import accuracy_score

score=accuracy_score(y_test, y_predict)
```
