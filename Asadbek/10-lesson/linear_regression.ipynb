```python
df=pd.read_csv("house.csv")
```

# Data review

```python
df.drop(columns=["ad_id", "description_block", "features", "no_of_units", "facility", "ber_classification", "surface"], inplace=True)
```

```python
df.info()
```

# Data preprocessing

# Handling missing values

```python
df.isnull().sum()
```

```python
for col in df.columns:
	if df[col].isnull().any():
		# if df[col].dtype == "object":
			df[col].fillna(df[col].mode()[0], inplace=True)
		# else: df[col].fillna(df[col].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df.tail()
```

# Encoding

```python
df.nunique()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()

for col in df.columns:
	if df[col].dtype == "object":
		if df[col].nunique()<=5:
			dummies=pd.get_dummies(df[col], prefix=col, dtype=int)
			df=pd.concat([df.drop(columns=[col]), dummies], axis=1)
		else: df[col]=encoder.fit_transform(df[col])
```

```python
df.head()
```

# Scaling

```python
df.info()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
scaler
```

```python
num_col=df.select_dtypes(include=["float64", "int64"]).columns.drop("price")
```

```python
num_col
```

```python
df[num_col]=scaler.fit_transform(df[num_col])
```

```python
df.head()
```

# Model selection

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()
```

```python
lr
```

# Training

```python
# input
x=df.drop("price", axis=1)
```

```python
# output
y=df["price"]
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
lr.fit(x_train, y_train)
```

# Prediction

```python
y_predict=lr.predict(x_test)
```

```python
y_predict[1]
```

```python
df.head()
```

# Evaluation

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
mae=mean_absolute_error(y_test, y_predict)
mse=mean_squared_error(y_test, y_predict)
r2=r2_score(y_test, y_predict)
```

```python
print("mae: ", mae)
print("mse: ", mse)
print("r2: ", r2)
```
