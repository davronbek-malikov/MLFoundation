```python
import pandas as pd

df=pd.read_csv("../airports.csv")
```

```python
df.isnull().sum()
```

```python
df.head()
```

# Drop columns

```python
columns = ["id", "icao", "iata", "gps_code", "local_code", "continent", "web_url", "wikipedia_url", "keywords", "timezone"]
df.drop(columns=columns, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.head()
```

```python
df.nunique()
```

# Mean

```python
# latitude

df["latitude"].fillna(df["latitude"].mean(), inplace=True)
```

# Mode

```python
# longitude
df["longitude"].fillna(df["longitude"].mode()[0], inplace=True)

# ident
df["ident"].fillna(df["ident"].mode()[0], inplace=True)

# country
df["country"].fillna(df["country"].mode()[0], inplace=True)

# region
df["region"].fillna(df["region"].mode()[0], inplace=True)

# municipality
df["municipality"].fillna(df["municipality"].mode()[0], inplace=True)

# scheduled_service
df["scheduled_service"].fillna(df["scheduled_service"].mode()[0], inplace=True)

# airport_type
df["airport_type"].fillna(df["airport_type"].mode()[0], inplace=True)
```

# Median

```python
# elevation

df["elevation"].fillna(df["elevation"].median(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.nunique()
```

# One-hot encoding

```python
# scheduled_service

dummies = pd.get_dummies(df["scheduled_service"], prefix="col", dtype=int)

dummies
```

```python
df=pd.concat([df.drop(columns=["scheduled_service"]), dummies], axis=1)
```

```python
df.info()
```

```python
df.nunique()
```

# Label encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
```

```python
df["ident"] = encoder.fit_transform(df["ident"])
df["name"] = encoder.fit_transform(df["name"])
df["country"] = encoder.fit_transform(df["country"])
df["region"] = encoder.fit_transform(df["region"])
df["municipality"] = encoder.fit_transform(df["municipality"])
df["airport_type"] = encoder.fit_transform(df["airport_type"])
```
