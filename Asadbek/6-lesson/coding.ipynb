# Agar berilgan son 5ga va 3ga bolinsa yaxshi aks holda yomon degan sozni ekranga chiqaring.

```python
num=15

if num % 5 == 0 and num % 3 == 0:
	print("Yaxshi")
else: print("Yomon")
```

# Agar berilgan sozning oxirgi harfi "a" bo'lsa yaxshi aks holda yomon degan sozni ekranga chiwaring

```python
str="Hello"

if str[-1] == "a":
	print("Yaxshi")
else: print("Yomon")
```

# Agar berilgan so'zning bosh harfi "b" va oxirgi harfi "a" bo'lsa yaxshi aks holda yomon degan so'zni ekranga chiwaring

```python
str="bolg'a"

if str[0] == "b" and str[-1] == "a":
	print("Yaxshi")
else: print("Yomon")
```

# Berilgan sonning musbat ekanini isbotlang

```python
num=10

if num>0:
	print("Berilgan son musbat")
elif num == 0:
	print("Berilgan son 0")
else: print("Berilgan son manfiy")
```
