# Keraksiz ustunni tashlab yuborish

```python
df.drop("id", axis=1, inplace=True)
```

# Mean (O'rta arifmetik)

```python
df["elevation"].fillna(df["elevation"].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df["ident"].fillna(df["ident"].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
df["longitude"].fillna(df["longitude"].median(), inplace=True)
```

```python
df["latitude"].fillna(df["latitude"].median(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df["icao"].fillna("AGGJ", inplace=True)
```

```python
df.isnull().sum()
```

# Drop (stun bo'yicha => 50% dan ko'p qiymatlr tushib qolgan bo'lsa ishlatamiz)

```python
df.drop("iata", axis=1, inplace=True)
df.drop("local_code", axis=1, inplace=True)
df.drop("continent", axis=1, inplace=True)
df.drop("web_url", axis=1, inplace=True)
df.drop("wikipedia_url", axis=1, inplace=True)
df.drop("keywords", axis=1, inplace=True)
df.drop("timezone", axis=1, inplace=True)
```

```python
df["gps_code"].fillna(df["gps_code"].mode()[0], inplace=True)
df["country"].fillna(df["country"].mode()[0], inplace=True)
df["region"].fillna(df["region"].mode()[0], inplace=True)
df["municipality"].fillna(df["municipality"].mode()[0], inplace=True)
df["scheduled_service"].fillna(df["scheduled_service"].mode()[0], inplace=True)
df["airport_type"].fillna(df["airport_type"].mode()[0], inplace=True)
```

# Discount Calculator


```python
price = 1200

if price > 1000:
    discount = price * 0.10
else:
    discount = 0

print("Final price:", price - discount)
```

# Largest of Three Numbers

```python
a = 10
b = 25
c = 15

if a >= b and a >= c:
    print("a is largest")
elif b >= a and b >= c:
    print("b is largest")
else:
    print("c is largest")
```
