# Print numbers


```python
# Print numbers from 1 to 10 using a for loop.

for num in range(1, 11):
	print(num)
```

# Sum of numbers

```python
# Find the sum of numbers from 1 to 100.

sum=0
for num in range(1, 101):
	sum = sum + num

print(sum)
```

# Even numbers

```python
# Print all even numbers between 1 and 50.

for num in range(1, 51):
	if num % 2 == 0:
		print(num)
```

# Count characters

```python
# Given a string, count how many characters it has using a for loop.

str="Hello"

sum=0
for i in str:
	sum = sum + 1

print(sum)
```

# Count vowels

```python
# Given a string, count how many vowels (a, e, i, o, u) it contains.

str="assalom"

count=0
for item in str:
	if item == "a" or item == "e" or item == "i" or item == "o" or item == "u":
		count = count + 1

print(count)
```

# Multiplication table

```python
# Print the multiplication table of 7 (7 × 1 to 7 × 10).

for num in range(1, 10):
	print(7 * num)
```

# Largest number

```python
# Given a list of numbers, find the largest number using a for loop

nums=[2, 5, 19, 1, 23, 49]

max=nums[0]

for num in nums:
	if max < num:
		max = num

print(max)
```

# Reverse a string

```python
# Reverse a string using a for loop.

text = "python"
reversed_text = ""

for ch in text:
    reversed_text = ch + reversed_text

print(reversed_text)
```

# Factorial

```python
# Given a number n, calculate its factorial using a for loop.

number=5
multiplication=1

for num in range(1, number + 1):
	multiplication = multiplication * num

print(multiplication)
```

# Count positive, negative, and zero

```python
# Given a list of numbers, count:
# how many are positive
# how many are negative
# how many are zero

numbers=[0, 3, 6, -6, 0, 2, -4, -10]
positive=0
negative=0
zero=0

for num in numbers:
	if num > 0:
		positive = positive + 1
	elif num < 0:
		negative = negative + 1
	else: zero = zero + 1

print("Positive:", positive)
print("Negative:", negative)
print("Zero:", zero)

```
