## 2-savol: O’yin o’ynay oladigan dastur yaratgan olim kim?

Alan Turing  
→ Sun’iy intellekt va kompyuter fanining asoschilaridan biri

---

## 3-savol: Machine learning qanday fan?

Machine learning — kompyuterga ma’lumotlar orqali  
o‘rganish va mustaqil qaror qabul qilishni o‘rgatadigan fan.

---

## 4-savol: Qaysi tadqiqot sizga qiziq bo‘ldi?

Masalan:  
Sun’iy intellektning inson kabi o‘rganishi.  
Chunki kompyuter tajriba orqali o‘rganishi juda qiziqarli  
va kelajak texnologiyasi hisoblanadi.

# Human Learning va Machine Learning

## 1-savol: Nima uchun MLdan oldin Human learningni o‘rganamiz?

Chunki ML inson o‘rganish jarayoniga o‘xshaydi.  
Inson tajriba orqali o‘rgansa, model ham data orqali o‘rganadi.

---

## 2-savol: Eng muhim o‘rganish turi qaysi?

Tajriba orqali o‘rganish.  
Chunki amaliyot orqali bilim mustahkam bo‘ladi  
va xatolardan o‘rganish mumkin.

---

## 3-savol: ML deganda nimani tushunasiz?

Kompyuterning ma’lumotlardan o‘rganib,  
bashorat qilish va qaror qabul qilish jarayoni.

---

## 4-savol: Jadvaldagi bosqichlardan eng muhimi qaysi?

Tajriba va amaliy mashq.  
Chunki o‘rganilgan bilimni qo‘llamasak  
natija bo‘lmaydi.

---

## 5-savol: Tajribani keyingi muammoga qo‘llashdagi asosiy xususiyat?

Generalization (umumlashtirish).  
Oldingi tajribadan yangi muammolarni hal qilish.

---

# Human Learning va Machine Learning qiyosi

## 1-savol: Ma’lumot ko‘payganda eng kerakli narsa nima?

To‘g‘ri va sifatli data.  
Qancha ko‘p va toza data bo‘lsa model shunchalik yaxshi o‘rganadi.

---

## 2-savol: Ma’lumotni ishlatishdan oldin nima qilish kerak?

- Data tozalash  
- Missing value to‘ldirish  
- Encoding  
- Scaling  

Chunki toza data bo‘lmasa model noto‘g‘ri o‘rganadi.

---

## 3-savol: Umumiy xulosa

Machine learning — data orqali o‘rganadigan tizim.  
To‘g‘ri data va preprocessing bo‘lsa model yaxshi ishlaydi.  
ML inson o‘rganishiga o‘xshaydi:  
tajriba → o‘rganish → natija → yaxshilash.
