```python
import pandas as pd

df = pd.read_csv("./../Data/bundesliga_player.csv")
df.head()
```

```python
df.isnull().sum() 
```

```python
# dtop 50% dan kop bosh joylar bolsa ochirish uchun ishlatiladi 
# coulumn gu 열 row 행 

df.drop(["full_name", "outfitter", "player_agent"], axis=1, inplace=True)
df.isnull().sum()
```

```python
df.info()
```

```python
# 필요없는 컬럼 삭제
df.drop("Unnamed: 0", axis=1, inplace=True)

# 문자 → mode
df["place_of_birth"].fillna(df["place_of_birth"].mode()[0], inplace=True)
df["foot"].fillna(df["foot"].mode()[0], inplace=True)
df["contract_expires"].fillna(df["contract_expires"].mode()[0], inplace=True)

# 가격 → median
df["price"].fillna(df["price"].median(), inplace=True)
df["max_price"].fillna(df["max_price"].median(), inplace=True)

# 확인
df.isnull().sum()
```

```python
df.info()
```

## Encoding asosan objectlar uchun ishledi Label encoding 

```python
# aynan bizga kerakli joyini chaqirib olamiz
from sklearn.preprocessing import LabelEncoder

# encoder ichiga solamiz

encoder = LabelEncoder()
# faqat objectlar uchun ishledi
"""
name
nationality
place_of_birth
position
foot
club
contract_expires
joined_club
"""
```

```python
df["name"] = encoder.fit_transform(df["name"])
df["nationality"] = encoder.fit_transform(df["nationality"])
df["place_of_birth"] = encoder.fit_transform(df["place_of_birth"])
df["position"] = encoder.fit_transform(df["position"])
df["foot"] = encoder.fit_transform(df["foot"])
df["club"] = encoder.fit_transform(df["club"])
df["contract_expires"] = encoder.fit_transform(df["contract_expires"])
df["joined_club"] = encoder.fit_transform(df["joined_club"])
```
