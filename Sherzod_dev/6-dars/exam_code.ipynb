# Mode

```python
# Mean contract_expires and player_agent  

df["contract_expires"].fillna(df["contract_expires"].mode()[0], inplace=True)
```

```python
df["foot"].fillna(df["foot"].mode()[0], inplace=True)
```

# Mean 


```python
df["max_price"].fillna(df["max_price"].mean(), inplace=True)
```

# Median

```python
df["price"].fillna(df["price"].median(), inplace=True)
```

# Fixed

```python
df["place_of_birth"].fillna("Uzbekistan", inplace=True)
```

```python
df.isnull().sum()
```

```python
df.drop(["Unnamed: 0", "full_name","player_agent", "outfitter"], axis=1, inplace=True)
df.isnull().sum()
```

# One-hot encoding 
##### harakteri bizning datamizga mos bolmaganligi uchun otkazib yubordik

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

df["name"] = label_encoder.fit_transform(df["name"])
df["nationality"] = label_encoder.fit_transform(df["nationality"])
df["place_of_birth"] = label_encoder.fit_transform(df["place_of_birth"])
df["position"] = label_encoder.fit_transform(df["position"])
df["foot"] = label_encoder.fit_transform(df["foot"])
df["club"] = label_encoder.fit_transform(df["club"])
df["contract_expires"] = label_encoder.fit_transform(df["contract_expires"])
df["joined_club"] = label_encoder.fit_transform(df["joined_club"])


df.info()
```
