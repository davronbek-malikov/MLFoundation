# 1. Supervised ML nima va qanday turlarga bo‘linadi?

AI/ML 3 turga bo‘linadi:
- Supervised ML
- Unsupervised ML
- Reinforcement ML

Supervised ML — input (features) va output (target) mavjud bo‘lgan o‘rganish turi.

Supervised ML 2 ga bo‘linadi:

1) Classification  
Target qiymat kategoriyalardan iborat bo‘ladi.

Classification turlari:
- Binary → targetda 2 ta class (ha/yo‘q, 0/1)
- Multi-class → targetda 3 va undan ortiq class

2) Regression  
Target qiymat sonli bo‘ladi va cheksiz qiymatga ega.

---

# 2. Data qanday bo‘lishi kerak?

Data 2 turga bo‘linadi:

1) Structured data  
Jadval ko‘rinishida bo‘ladi (qator va ustunlar).

2) Unstructured data  
Rasm, video, audio, text kabi formatlar.

Data sifatli bo‘lishi kerak:
- Ishonchli
- Yetarlicha ko‘p
- Toza
- Mavzuga mos

---

# 3. Data preprocessing nima?

ML model yaratishdagi muhim bosqich.  
Bu bosqichda data tozalanadi va model tushunadigan ko‘rinishga keltiriladi.

Asosiy 3 bosqich:
- Missing values bilan ishlash
- Encoding (categorical → numerical)
- Scaling (qiymatlarni balanslash)

---

# 4. Mean va Mode farqi

Mean:
- O‘rta arifmetik qiymat
- Faqat numerical ustunlar uchun

Mode:
- Eng ko‘p takrorlangan qiymat
- Numerical va categorical uchun ishlaydi

Ikkalasi ham missing value to‘ldirishda ishlatiladi.

---

# 5. Regression va Classification

Regression:
Target sonli va cheksiz qiymatga ega.

Misol:
Uy narxini predict qilish  
Target → narx

Classification:
Target kategoriyali va chekli.

Misol:
Yomg‘ir yog‘adi / yog‘maydi  
Target → 2 class (binary)

---

# 6. Missing value to‘ldirish usullari

5 ta usul mavjud:

1. Mean  
→ o‘rtacha qiymat (numerical)

2. Mode  
→ eng ko‘p takrorlangan (numerical + categorical)

3. Median  
→ o‘rtadagi qiymat (numerical)

4. Fixed  
→ o‘zimiz qiymat beramiz

5. Drop  
→ qatorni o‘chirib tashlaymiz

---

# 7. ML va AI da eng muhim element

Eng muhim element — data.

Sababi:
Model data orqali o‘rganadi.  
Data sifati yaxshi bo‘lsa model ham yaxshi ishlaydi.

---

# 8. Median qanday ishlaydi?

Median — o‘rtadagi qiymat.

Bosqichlar:
1. Sonlar kichikdan kattaga saralanadi
2. O‘rtadagi son olinadi

Agar sonlar soni toq bo‘lsa:
→ o‘rtadagi son median

Agar juft bo‘lsa:
→ o‘rtadagi 2 son o‘rtacha qiymati olinadi
