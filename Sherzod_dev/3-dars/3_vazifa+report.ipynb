## AI (Sun’iy idrok) — 인공지능

### Supervised Machine Learning (지도학습)

Supervised ML — bu input (X) va output (Y) bilan o‘qitiladigan AI.

- X → kirish ma’lumot  
- Y → javob (label, natija)

Ya’ni:  
Biri savol bo‘lsa → ikkinchisi javob

Model vazifasi:  
X ni ko‘rib Y ni to‘g‘ri topishni o‘rganish

---

### Misol

Uyning kattaligi → Narxi

- Uy kattaligi = X  
- Uy narxi = Y  

Model o‘rganadi:  
Agar uy katta bo‘lsa → narx qimmat bo‘ladi

---

## Supervised ML turlari

### 1. Classification (분류)

Natija kategoriya bo‘ladi.

#### Binary Classification (2 ta natija)

Faqat 2 ta javob:

- Yes / No  
- True / False  
- Spam / Normal  
- Mushuk / It  

2 ta class mavjud.

---

#### Multi-class Classification (3 ta yoki ko‘p)

3 ta yoki undan ko‘p javob:

- Olma / Banan / Apelsin  
- Mushuk / It / Qush  
- 멧돼지 / 고라니 / 너구리  

3 yoki undan ko‘p class mavjud.

---

### 2. Regression (회귀)

Natija son bo‘ladi.

Misol:
- Uy narxi (350000$)
- Harorat (27°C)
- Sotuv miqdori

Natija raqam (continuous value).

---

## Qisqa xulosa

| Tur | Natija |
|-----|-------|
| Classification | Kategoriya |
| Binary | 2 ta javob |
| Multi-class | 3+ javob |
| Regression | Raqam |

Supervised ML =  
X (input) → model → Y (output)
