# Machine Learning asoslari

## AI → ML → Turlari

Artificial Intelligence (AI)  
→ Machine Learning (ML)  
→ Supervised Machine Learning (SML)  
→ Unsupervised Machine Learning (USML)  
→ Reinforcement Machine Learning (RML)

---

# Supervised Machine Learning (SML)

Supervised ML — input (X) va output (Y) mavjud bo‘lgan o‘rganish turi.  
Modelga savol va javob (label) beriladi va model shu orqali o‘rganadi.

## Supervised ML 2 ga bo‘linadi:

### 1. Classification
Ma’lumotni kategoriyalarga ajratish.

#### Binary classification
Target (Y) faqat 2 xil qiymatga ega bo‘ladi.

Misol:
- Ha / Yo‘q
- Spam / Spam emas
- Kasal / Sog‘lom

#### Multi-class classification
Target (Y) 3 yoki undan ortiq qiymatga ega.

Misol:
- Qizil / Ko‘k / Yashil
- Mushuk / It / Qush
- A / B / C baho

---

### 2. Regression
Natija son (raqam) bo‘ladi.  
Qiymatlar cheksiz bo‘lishi mumkin.

Misol:
- Uy narxi
- Temperatura
- Mashina tezligi
- Oylik daromad

---

# Unsupervised Machine Learning (USML)

Bu usulda target (javob) yo‘q.  
Model faqat ma’lumotni o‘zi tahlil qiladi.

Misol:
- Customer segmentation
- O‘xshash ma’lumotlarni guruhlash

---

# Reinforcement Machine Learning (RML)

Model harakat qiladi va mukofot (reward) orqali o‘rganadi.

Misol:
- Robot yurishi
- O‘yin o‘ynaydigan AI
- Avtonom mashina

---

# ML model yaratish bosqichlari

Model yaratish quyidagi ketma-ketlikda amalga oshiriladi:

1. Data collection  
   Dataset yig‘ish (Kaggle, Github va boshqalar)

2. Data preprocessing  
   Tozalash, encoding, scaling

3. Algorithm tanlash  
   Logistic Regression, RandomForest va boshqalar

4. Model training  
   Modelni o‘qitish

5. Evaluation  
   Modelni baholash (accuracy va boshqalar)

6. Testing  
   Yangi data bilan test qilish

7. Deployment  
   Modelni real loyihaga joylash

8. Monitoring  
   Model ishlashini kuzatish

# If-else condition (shart operatori)

Ma’lumotni tekshirish va natija chiqarish uchun ishlatiladi.

Misol:  
25 ta test savoli bor.

Agar:
- 23 dan katta → 5 baho
- 18–23 → 4 baho
- 18 dan kichik → 3 baho

Shart logikasi:  
Agar natija katta bo‘lsa → yuqori baho  
aks holda → pastroq baho

If-else yordamida model yoki data tekshirish mumkin.
