# ML Algorithm Family

Machine Learning algoritmlari 2 ta asosiy guruhga bo‘linadi:

- Classification (kategoriya bashorat qilish)
- Regression (son bashorat qilish)

---

# Linear va Non-linear algoritmlar

## Linear algoritmlar

### Linear Regression (Regression)
Sonli natijani bashorat qiladi.

Misol:
- Uy narxi
- Oylik maosh
- Temperatura

### Logistic Regression (Classification)
Kategoriya bashorat qiladi.  
Ha/yo‘q, 0/1 kabi natijalar.

Sodda va kichik datasetlar uchun juda yaxshi ishlaydi.

---

## Non-linear algoritmlar

### 1. Tree-based algoritmlar

### 2. Distance-based algoritmlar

# Logistic Regression

Classification tasklar uchun ishlatiladi.  
Sodda, tez va kichik datasetlar uchun mos.

---

# Logistic Regression bosqichlari

1. Data preprocessing  
   → cleaning, encoding, scaling

2. X va Y aniqlash  
   → input va target

3. Splitting  
   → train va testga ajratish

4. Training (fitting)  
   → modelni o‘qitish

5. Predict  
   → natija bashorat qilish

6. Evaluation  
   → model aniqligini tekshirish

---

# Coding bosqichlari

## 1. Input va output ajratish

"x = df.drop('target_col_name', axis=1)"  
→ input (features)

"y = df['target_col_name']"  
→ target (output)

---

## 2. Train va testga ajratish

"from sklearn.model_selection import train_test_split"

"x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)"

test_size=0.2  
→ 20% test, 80% train

random_state=42  
→ har safar bir xil bo‘linadi

"x_train.shape"  
→ train data shaklini ko‘rish

---

## 3. Logistic Regression import qilish

"from sklearn.linear_model import LogisticRegression"

"log_reg = LogisticRegression()"  
→ model yaratildi

---

## 4. Training (fitting)

"log_reg.fit(x_train, y_train)"

Model train data orqali o‘rganadi.

---

## 5. Predict qilish

"y_predict = log_reg.predict(x_test)"

Test data uchun bashorat qiladi.

---

## 6. Evaluation (baholash)

"from sklearn.metrics import accuracy_score"

"accuracy_score(y_test, y_predict)"

Natija:
0 va 1 oralig‘ida qiymat chiqadi.

Misol:
0.92 → 92% to‘g‘ri bashorat  
0.75 → 75% to‘g‘ri bashorat

Bu model aniqligi (accuracy) hisoblanadi.
