```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
from sklearn.preprocessing import LabelEncoder

encoder=LabelEncoder()
standardScaler = StandardScaler()
minMaxScaler = MinMaxScaler()
robustScaler = RobustScaler()

df = pd.read_csv("./../Data/bundesliga_player.csv")
```

```python
df.drop(["full_name","contract_expires","outfitter","player_agent"], axis=1, inplace=True)
df.isnull().sum()
```

```python
columns=df.columns

for col in columns:
    if df[col].isnull().any():
        if df[col].dtype == "object":
            df[col].fillna(df[col].mode()[0], inplace=True)
        else: df[col].fillna(df[col].median(), inplace=True)


df.isnull().sum()
```

```python
df.info()
```

```python
columns=df.columns

for col in columns:
    if df[col].dtype == "object":
        if df[col].nunique() <= 2:
            dummies=pd.get_dummies(df[col], prefix="col", dtype=int)
            df=pd.concat([df.drop(columns=[col]), dummies], axis=1)
        else: df[col]=encoder.fit_transform(df[col])

df.info()


```

```python
nummerical_col = df.select_dtypes(include=['int64','float64']).columns
for col in nummerical_col:
    df[col]=robustScaler.fit_transform(df[[col]])
    
    
    
    
    
```
