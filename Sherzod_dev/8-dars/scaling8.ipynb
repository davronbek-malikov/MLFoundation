```python
import pandas as pd

df = pd.read_csv("./../Data/bundesliga_player.csv")
df.info() 
df.isnull().sum()
```

```python
df.drop(["full_name","contract_expires","player_agent","outfitter"], axis=1, inplace=True)

df.isnull().sum()
```

```python
for col in df.columns:
    if df[col].isnull().any():
        if df[col].dtype == "object":
            df[col].fillna(df[col].mode()[0], inplace=True)
        else: df[col].fillna(df[col].median(), inplace=True)

df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()

columns=df.columns

for col in columns:
    if df[col].dtype == "object":
        if df[col].nunique() <= 2:
            dummies=pd.get_dummies(df[col], prefix="col", dtype=int)
            df=pd.concat([df.drop(columns=[col]), dummies], axis=1)
        else: df[col]=encoder.fit_transform(df[col])

df.info()

```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler

standart_scaler = StandardScaler()
minmax_scaler = MinMaxScaler()
robust_scaler = RobustScaler()

num_col = df.select_dtypes(include= ['int64','float64'])
for col in num_col:
    df[col] = robust_scaler.fit_transform(df[[col]])
df.head()
# ÎÅù
```
