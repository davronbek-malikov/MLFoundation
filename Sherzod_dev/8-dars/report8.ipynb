# Scaling (Ma’lumotlarni masshtablash)

Scaling — juda katta va juda kichik sonli qiymatlarni  
bir xil diapazonga keltirish jarayoni.

Model katta sonlarni kichik sonlardan ustun deb o‘ylamasligi uchun  
qiymatlarni balanslashtiramiz.

---

# Scaling qachon kerak?

Scaling qo‘llash uchun:

1. Ustun numeric (sonli) bo‘lishi kerak  
2. Agar qiymatlar allaqachon 0–1 oralig‘ida bo‘lsa → scaling shart emas  
3. Target (Y) ustuniga scaling qilinmaydi

---

# Scaling usullari

Scalingning 3 ta asosiy usuli bor:

- Standard Scaling
- MinMax Scaling
- Robust Scaling

---

## 1. MinMax Scaling

Qiymatlarni 0 va 1 oralig‘iga olib keladi.

Formula:
(x − min) / (max − min)

Natija:
- Eng kichik qiymat → 0
- Eng katta qiymat → 1
- Qolganlari → 0 va 1 orasida

"scaler = MinMaxScaler()"  
"df['col'] = scaler.fit_transform(df[['col']])"

Bu usulda manfiy son chiqmaydi.

---

## 2. Standard Scaling

Qiymatlarni standart normal taqsimotga keltiradi.  
Mean = 0, Standard deviation = 1

Natijada:
- Musbat ham chiqadi
- Manfiy ham chiqadi

"scaler = StandardScaler()"  
"df['col'] = scaler.fit_transform(df[['col']])"

Bu usul ML modellarda eng ko‘p ishlatiladi.

---

## 3. Robust Scaling

Outlier (juda katta yoki kichik qiymatlar) bo‘lsa  
eng yaxshi usul.

Median va IQR asosida hisoblanadi.

Natijada:
- Manfiy qiymatlar chiqishi mumkin
- Outlier ta’siri kamayadi

"scaler = RobustScaler()"  
"df['col'] = scaler.fit_transform(df[['col']])"

# Numerical ustunlarni aniqlash

Faqat sonli ustunlarga scaling qilinadi.

"numCol = df.select_dtypes(include=['int64','float64']).columns.drop('target_col_name')"

- numeric ustunlarni topadi
- target ustunni olib tashlaydi

# Bitta ustunga scaling qo‘llash

Standard scaling:

"scaler = StandardScaler()"  
"df['colName'] = scaler.fit_transform(df[['colName']])"

MinMax scaling:

"scaler = MinMaxScaler()"  
"df['colName'] = scaler.fit_transform(df[['colName']])"

Robust scaling:

"scaler = RobustScaler()"  
"df['colName'] = scaler.fit_transform(df[['colName']])"

---

# For loop orqali scaling (avtomatik)

## 1-usul (objectdan tashqari hammasiga)

"scaler = MinMaxScaler()"  

"for col in df.columns:"  
"    if df[col].dtype != 'object':"  
"        df[col] = scaler.fit_transform(df[[col]])"

Bu usul:
- string ustunlardan tashqari
- hamma numeric ustunlarga scaling qiladi

---

## 2-usul (faqat numeric ustunlar)

"num_col = df.select_dtypes(include=['int64','float64']).columns.drop('target_colName')"  

"for col in num_col:"  
"    df[col] = scaler.fit_transform(df[[col]])"

Bu usul:
- faqat numeric ustunlarga
- targetdan tashqari scaling qiladi
