# Data Processing:Review

```python
import pandas as pd

df = pd.read_csv("./../Data/bundesliga_player.csv")
df.head()
```

```python
df.isnull().sum()
# datada name bor full nameda muammo 50% dan kop 
```

```python
df.drop(["full_name","outfitter","player_agent"],axis=1,inplace=True)
df.isnull().sum()
```

```python
columns=df.columns

for col in columns:
    if df[col].isnull().any():
        if df[col].dtype == "object":
            df[col].fillna(df[col].mode()[0],inplace=True)
        else:
            df[col].fillna(df[col].median(),inplace=True)


df.info()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()


for col in df.columns: # dataning columini olamiz 
    if df[col].dtype == "object": # agar object bo'lsa 
        df[col] = label_encoder.fit_transform(df[col]) # va raqamlashtramiz

df.info()
```
