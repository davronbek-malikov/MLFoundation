# Encoding va For Loop tushunchasi

## Encoding nima?

Encoding — kategoriyali (matnli) ma’lumotlarni  
model tushunadigan sonlarga aylantirish jarayoni.

Model matnni tushunmaydi, faqat sonlarni tushunadi.  
Shuning uchun kategoriyalarni raqamga o‘zgartiramiz.

Encoding turlari:
- One-hot encoding
- Label encoding
- Frequency encoding
- Target encoding
- Ordinal encoding

---

# 1. One-Hot Encoding

Har bir kategoriya alohida ustunga bo‘linadi.  
0 va 1 orqali belgilanadi.

Misol:
airport_type → small, medium, large

Natija:
small → 1 0 0  
medium → 0 1 0  
large → 0 0 1  

"pd.get_dummies(df['airport_type'])"

Bu usulda kategoriyalar orasida tartib (ranking) yo‘q.

---

# 2. Label Encoding

Har bir kategoriya bitta son bilan almashtiriladi.

small → 0  
medium → 1  
large → 2  

Bu usul oddiy, lekin model noto‘g‘ri tartib deb o‘ylashi mumkin.

---

# 3. Frequency Encoding

Kategoriya necha marta takrorlanganini  
umumiy elementlar soniga bo‘lib chiqiladi.

Formula:
Frequency = kategoriya soni / umumiy son

Misol:

airport_type ustuni:
small  
large  
small  
medium  
medium  
small  

Hisob:
small → 3/6 = 0.5  
large → 1/6  
medium → 2/6  

Natija:

airport_type → airport_type_freq  
small → 0.5  
large → 1/6  
small → 0.5  
medium → 2/6  
medium → 2/6  
small → 0.5  

Bu usulda kategoriya nechta ekanligi modelga ko‘rinadi.

---

# 4. Target Encoding (Mean encoding)

Kategoriya bilan bog‘liq target qiymatlar o‘rtachasi olinadi.

Har bir kategoriya uchun:
o‘sha kategoriya → target o‘rtacha qiymati

Misol:

airport_type      target(Good)  
small             0  
large             1  
small             0  
medium            1  
medium            1  
small             1  

Hisob:

small → (0+0+1)/3 = 1/3  
large → 1/1 = 1  
medium → (1+1)/2 = 1  

Natija:

airport_type → airport_type_target  
small → 1/3  
large → 1  
small → 1/3  
medium → 1  
medium → 1  
small → 1/3  

Bu encoding modelga kategoriya qanchalik yaxshi ekanini ko‘rsatadi.

---

# 5. Ordinal Encoding

Kategoriya tartibga ega bo‘lsa ishlatiladi.  
Ketma-ketlik asosida son beriladi.

Misol:
small < medium < large

small → 1  
medium → 2  
large → 3  

Natija:

airport_type → airport_type_ordinal  
small → 1  
large → 3  
small → 1  
medium → 2  
medium → 2  
small → 1  

Bu usulda tartib muhim bo‘lsa ishlatiladi.

---

# For Loop tushunchasi

For loop — takrorlanuvchi jarayonlarni  
avtomatlashtirish uchun ishlatiladi.

Bir xil kodni ko‘p marta yozmaslik uchun ishlatiladi.

Misol:
Bizda list bor:

"str = ['small','large','large','medium']"

Har bir elementni chiqarish:

Agar oddiy yozsak:
"print(str[0])"  
"print(str[1])"  
"print(str[2])"  
"print(str[3])"  

Bu juda uzun va noqulay.

For loop orqali avtomatik bajariladi:

"for i in str:
    print(i)"

Bu yerda:
for → takrorlash  
i → har bir element  
str → list

Loop har bir elementni bittadan chiqaradi.

---

# Qisqa xulosa

Encoding  
→ matnni sonlarga aylantirish

One-hot  
→ har kategoriya alohida ustun

Label  
→ har kategoriya bitta son

Frequency  
→ takrorlanish ulushi

Target  
→ target o‘rtacha qiymati

Ordinal  
→ tartib asosida son berish

For loop  
→ takrorlanuvchi jarayonni avtomatlashtirish
