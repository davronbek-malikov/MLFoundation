Machine Learning 전체 구조 (ML Pipeline)

ML Model Structure

1. Data Collection  
   Raw data 수집  

2. Data Preprocessing  
   데이터 정리 및 변환  
   (결측치 처리, 인코딩, 스케일링)

3. Model Selection  
   알고리즘 선택  

4. Training (Fitting)  
   모델 학습  

5. Prediction  
   새로운 데이터 입력 → 결과 출력  

6. Evaluation  
   성능 평가 (Accuracy 등)  

7. Testing  
   실제 테스트  

8. Deployment  
   서비스/실전 적용  


Data Preprocessing

1. Handling Missing Values (결측치 처리)

MEAN → 숫자형 평균값으로 채움  
MEDIAN → 중앙값 (outlier 있을 때 좋음)  
MODE → 가장 많이 나온 값  
DROP → 행/열 삭제  
FIXED → 직접 값 입력  


2. Encoding (문자 → 숫자 변환)

Categorical 데이터를 숫자로 변환

One-hot encoding → 클래스 수만큼 컬럼 생성  
Label encoding → 알파벳 순서대로 숫자 변환  
Frequency encoding → 등장 비율로 변환  
Target encoding → target 평균값  
Ordinal encoding → 순서 기반 숫자  


3. Scaling (스케일링)

값 크기 균형 맞추기

Standard → 평균 0, 표준편차 1  
MinMax → 0~1 범위  
Robust → 이상치(outlier)에 강함  

MinMax → 0~1 사이로 변환

Linear Regression (회귀)

공식  
y = B0 + B1x + e  

x → input  
y → output  
B → 파라미터  
e → 오차 (항상 존재)

종류  
- Simple Linear Regression (변수 1개)  
- Multiple Linear Regression (변수 여러 개)
