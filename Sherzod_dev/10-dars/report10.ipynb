# ML umumiy jarayoni (o‘qituvchi tushuntirgan metod)

Raw data → Data preprocessing → Algorithm selection  
Bu mashinaviy o‘rganish modelini yaratishning asosiy ketma-ketligi.

---

# ML Model Structure

1. Data collection (Raw data)  
   → Dataset yig‘ish (Kaggle, Github va boshqalar)

2. Data preprocessing  
   → Missing value bilan ishlash  
   → Encoding  
   → Scaling / normalization  

3. Model (Algorithm) selection  
   → Qaysi model ishlatiladi (LR, DT, RF va boshqalar)

4. Model training (fitting)  
   → Modelni train data orqali o‘qitish

5. Prediction  
   → Model yangi data kirganda natija qaytaradi

6. Evaluation  
   → Model qanchalik yaxshi ishlayotganini tekshirish  
   → Agar accuracy past bo‘lsa model improvement qilish kerak

7. Testing  
   → Real natija olish

8. Deployment  
   → Modelni real loyiha yoki tizimga joylash

---

# Linear Regression

Linear regression:
- Linear family
- Baseline model
- Regression algorithm

Asosan sonli qiymatlarni bashorat qilish uchun ishlatiladi.

Misol:
- Uy narxi
- Maosh
- Savdo miqdori

---

# Linear Regression maqsadi

Input va output orasidagi eng mos chiziqni topish.

Model eng yaxshi chiziqni topadi va  
shu orqali bashorat qiladi.

---

# Linear Regression formulasi

y = B0 + B1x + e

y → output (natija)  
x → input (feature)  
B (beta) → parametrlar  
e (epsilon) → error  

Error hech qachon 0 bo‘lmaydi.  
e ≠ 0

Model doim kichik xatolik bilan ishlaydi.

---

# Kuchli tomonlari

# Evaluation (model baholash)

Model qanchalik yaxshi ishlayotganini tekshiradi.

Savollar:
1. Prediction qanchalik to‘g‘ri?  
2. Model ishonchlimi?  
3. Generalization bormi?  
4. Overfitting bormi?  

---

# Classification evaluation

Classification modellarda:

- Accuracy score
- Classification report

Accuracy → nechta to‘g‘ri bashorat qilganini ko‘rsatadi

---

# Regression evaluation

Regression modellarda quyidagilar ishlatiladi:

1. MAE (Mean Absolute Error)  
2. MSE (Mean Squared Error)  
3. RMSE (Root Mean Squared Error)  
4. R² score

---

## MAE, MSE, RMSE

Bu ko‘rsatkichlar errorni o‘lchaydi.

Error:
original qiymat − predict qiymat

Farq qanchalik kichik bo‘lsa  
model shunchalik yaxshi.

### MAE
O‘rtacha xatolik

### MSE
Xatolik kvadratlari o‘rtachasi

### RMSE
Kvadrat ildizli xatolik

Qiymatlar kichik bo‘lishi kerak.

---

## R² score (coefficient of determination)

Model qanchalik yaxshi tushuntirayotganini ko‘rsatadi.

0 ga yaqin → yomon model  
1 ga yaqin → yaxshi model  

---

# Qisqa xulosa

ML jarayoni:
raw data → preprocessing → model → training → evaluation → deploy

Linear regression:
→ regression uchun
→ eng sodda model
→ baseline model

Evaluation:
classification → accuracy  
regression → MAE, MSE, RMSE, R²  

Error kichik bo‘lsa → model yaxshi


숫자 예측 → Linear Regression (기본), RandomForest/XGBoost (정확도 필요)  
→ 결과가 가격·매출처럼 숫자일 때 사용하는 모델

분류 문제 → Logistic Regression (기본), RandomForest (실무), XGBoost (고성능)  
→ 결과가 합격/불합격, 스팸/정상처럼 종류일 때 사용

이미지 데이터 → CNN (이미지 인식), YOLO (객체 탐지)  
→ 사진이나 영상에서 물체를 찾거나 분류할 때 사용

텍스트 데이터 → LLM (GPT, BERT 등 자연어 처리)  
→ 챗봇, 번역, 문서요약 등 문장 이해·생성할 때 사용

추천 시스템 → RandomForest 또는 Vector Search + LLM  
→ 사용자 취향 분석해서 상품·음식·영상 추천할 때 사용

baseline 모델 → Linear Regression / Logistic Regression 먼저 사용  
→ 가장 기본 모델로 먼저 성능 기준(baseline)을 확인할 때 사용

정확도 높이기 → RandomForest → XGBoost 순서  
→ 기본 모델보다 더 높은 성능이 필요할 때 단계적으로 사용

딥러닝 사용 → 이미지(CNN), 텍스트(LLM)일 때  
→ 데이터가 이미지·음성·텍스트처럼 복잡할 때 사용
