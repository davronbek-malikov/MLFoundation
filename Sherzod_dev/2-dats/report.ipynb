# Data Types (Ma’lumot turlari)

## Data typelar nima uchun kerak?

Data type — ma’lumot turini bildiradi.  
Kompyuterga ma’lumot kiritishdan oldin va  
coding jarayonida ma’lumotlar turini bilish muhim.

Data type:
- ma’lumotni to‘g‘ri saqlash
- hisoblashni to‘g‘ri bajarish
- xatoliklarning oldini olish  
uchun kerak bo‘ladi.

---

# Asosiy data typelar

## int (integer)
Butun sonlar.

Misol:
- 5  
- 100  
- -3  

"age = 25"

---

## float
O‘nlik (kasr) sonlar.

Misol:
- 3.14  
- 10.5  
- -2.7  

"price = 10.5"

---

## str (string)
Matn yoki so‘zlar.

Misol:
- "Hello"
- "Machine learning"
- "Seoul"

"name = 'Ali'"

---

## boolean (bool)
Faqat 2 ta qiymatga ega:

- True
- False

Misol:
"passed = True"  
"online = False"

Shart tekshirishda ishlatiladi.

---

# Data typelar qayerda ishlatiladi?

- Coding jarayonida
- Dataset bilan ishlaganda
- Hisoblashlarda
- Shart (if-else) tekshirishda
- Machine learning preprocessingda

---

# Index tushunchasi

Index — element tartib raqami.

List yoki string ichidagi har bir element  
index orqali olinadi.

Index har doim 0 dan boshlanadi.

Misol:
"word = 'python'"

p → index 0  
y → index 1  
t → index 2  

"word[0]" → p  
"word[3]" → h

---

# Index operations

Elementlarni index orqali olish:

"list = [10,20,30,40]"

"list[0]" → 10  
"list[2]" → 30  

Oxiridan olish:

"list[-1]" → oxirgi element  
"list[-2]" → oxiridan ikkinchi

---

# Qisqa xulosa

Data type  
→ ma’lumot turini bildiradi

Asosiy turlar:
- int → butun son
- float → kasr son
- str → matn
- bool → True/False

Index  
→ element tartib raqami  
→ 0 dan boshlanadi
