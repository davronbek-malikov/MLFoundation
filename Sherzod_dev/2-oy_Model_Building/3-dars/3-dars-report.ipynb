# Regression Model Evaluation Metrics

Mashinali o‘rganishda (**Machine Learning**) regression modellarning aniqligini baholash uchun quyidagi metrikalar ishlatiladi.

---

# 1. MAE (Mean Absolute Error)
**Ta'rif:**  
MAE — haqiqiy qiymat (`y`) va model bashorati (`ŷ`) orasidagi **o‘rtacha mutloq farq**.

### Formula

$$
MAE = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i|
$$

### Ishlash mantig‘i

1. Har bir nuqtadagi xato hisoblanadi  
2. Xatoning **moduli** olinadi  
3. Barcha xatolar yig‘iladi  
4. Ma'lumotlar soniga bo‘linadi  

---

# 2. MSE (Mean Squared Error)

**Ta'rif:**  
MSE — haqiqiy qiymat va bashorat orasidagi farqning **kvadrati o‘rtachasi**.

### Formula

$$
MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

### Ishlash mantig‘i

1. Bashorat va haqiqiy qiymat ayirmasi olinadi  
2. Farq **kvadratga oshiriladi**  
3. Barcha kvadrat xatolar yig‘iladi  
4. O‘rtacha qiymat olinadi  

---

# 3. RMSE (Root Mean Squared Error)

**Ta'rif:**  
RMSE — bu **MSE ning kvadrat ildizi**.

### Formula

$$
RMSE = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2}
$$

### Ishlash mantig‘i

1. Avval **MSE hisoblanadi**  
2. Natijadan **kvadrat ildiz olinadi**

---

# 4. R² Score (Coefficient of Determination)

**Ta'rif:**  
R² modelning ma'lumotlarni qanchalik yaxshi tushuntira olishini ko‘rsatadi.

### Formula

$$
R^2 = 1 - \frac{SS_{res}}{SS_{tot}}
$$

### Bu yerda

**Residual Sum of Squares**

$$
SS_{res} = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

**Total Sum of Squares**

$$
SS_{tot} = \sum_{i=1}^{n} (y_i - \bar{y})^2
$$

---
