# Funksiyaga oid vazifa


```python
def musbat_manfiy(son):
    if son>0:
        return "musbat"
    elif son <0:
        return "manfiy"
    else:
        return "0"
    
musbat_manfiy(44)
```

```python
def toq_juft(son):
    if son % 2 == 0:
        return "juft"
    else:
        return "toq"
    
toq_juft(11)
```

```python
def son_qosh(a,b):
    return a + b
son_qosh(2,3)
```

```python
def sonlar_royhati(sonlar):
    jami =0
    for son in sonlar:
        jami = jami + son
    return jami
list_son = [10,20,30,30]
print(sonlar_royhati(list_son))
```
