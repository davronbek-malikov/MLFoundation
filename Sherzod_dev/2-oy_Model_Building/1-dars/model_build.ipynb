```python
import pandas as pd 
df = pd.read_csv("/Users/sherzodbeksharobiddinov/MLFoundation/Sherzod_dev/2-oy_Model_Building/Data/bank-full.csv")
df.info()
```

```python
def toldir(df):
    for col in df.columns:
        if df[col].isnull().any():
            if df[col].dtype == 'object':
                df[col].fillna(df[col].mode()[0], inplace=True)
            else:
                df[col].fillna(df[col].mean(), inplace=True)
    return df
df = toldir(df)

df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

def encodela(df):
    encodla = LabelEncoder()
    for col in df.columns:
        if df[col].dtype == "object":
            df[col] = encodla.fit_transform(df[col])
    return df
df = encodela(df)
df.info()
```

```python
num_col = df.select_dtypes(include = ['int64','float64']).columns.drop('y')
print(len(num_col))
```

```python
from sklearn.preprocessing import MinMaxScaler
def scalerla(df,num_col):
    scaler = MinMaxScaler()
    df[num_col]=scaler.fit_transform(df[num_col])
    return df
num_col = df.select_dtypes(include = ['int64','float64']).columns.drop('y')
scalerla(df,num_col)
```

```python
x = df.drop('y',axis=1)
y=df['y'].astype(int)
print(len(x))
print(len(y))
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(
    x,y, test_size = 0.2,random_state = 42
)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()

lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
y_pred[1:10]
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"mae: {mae}")
print(f"mse: {mse}")
print(f"r2_score: {r2}")
```
