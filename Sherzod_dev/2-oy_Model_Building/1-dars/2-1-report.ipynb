# Machine Learning: Asosiy Tushunchalari

## 1. Algorithm (Algoritm)

### Ta'rif
**Algoritm** — bu modelni qurish uchun ishlatiladigan yo‘riqnomalar to‘plami yoki matematik qoidalar.

### Vazifasi
Ma'lumotlardagi qonuniyatlarni o‘rganish.

### O‘rni
Algoritm model tayyor bo‘lgunga qadar, ya'ni **"o‘rganish" (training) jarayonida** faol ishlaydi.

---

## 2. Model

### Ta'rif
**Model** — bu algoritm ma'lumotlarni o‘rganib bo‘lgandan keyin hosil bo‘lgan **yakuniy natija (bilim)**.

### Amaliyoti
Biz modelni real hayotda (**production environment**) quyidagi vazifalar uchun ishlatamiz:

- Yangi ma'lumotlarni **bashorat qilish (prediction)**
- Ma'lumotlarni **tahlil qilish (analysis)**

---

## 3. Model Evaluation (Modelni Baholash)

### Ta'rif
**Model Evaluation** — model qanchalik aniq ishlayotganini tekshirish jarayoni.

### Asosiy metrikalar

#### R² (R-squared)
Model ma'lumotlarga qanchalik mos kelishini ko‘rsatadi.

#### MSE (Mean Squared Error)
Xatolik kvadratlarining o‘rtacha qiymati.

#### RMSE (Root Mean Squared Error)
MSE ning kvadrat ildizi. Model xatoligini real birlikda ko‘rsatadi.

#### MAE (Mean Absolute Error)
Xatoliklarning absolut o‘rtacha qiymati.

> Xatolik qancha **kam** bo‘lsa, model shuncha **yaxshi** hisoblanadi.
