biz shu paytgacha asoslan tran_test_splitdan foydalandik bunda asosa 0.2 yoki 0.3 yani 20% testga 80% taninga yani 100 tadatamiz bor 20 tasi test 80 tasi train uchun lekin bunda bitta muammo bor 80% aniq data bolib 20% no aniq bolsachi?

K-Fold (K-bo'lakka bo'lish) esa bunga barham beradi! KFold ma'lumotlarni nafaqat bir marta o'qiydi va test qiladi, balki buni navbati bilan K - marta takrorlaydi!

Dasturda siz: kf = KFold(n_splits = 5, shuffle=True, random_state=42) yozgan edingiz. Tahlili:

n_splits = 5: Kitobdagi jami 100% savollarni teng 5 ta blokga bo'lib (har birida 20% dan) guruhlaymiz.
Shunda mantiq: Birinchi bosqichda kompyuter 1, 2, 3 va 4-blokni o'qiydi (train), hamda 5-blokda imtihon topshiradi (test). Ikkinchi bosqichda endi o'rganilgan hamma narsani ununtib boshqattan model tuzadi-da, faqat 1, 2, 3 va 5-bloklarni yodlab (train), endi navbatdagi 4-bloktegi ma'lumotlardan imtihon (test) topshiradi... Shu tarzda umumiy ma'lumot hamma blokda test topshirib chiqish kafolatlanadi! Qolib ketgan hech qanday "ko'rilmagan holat" bo'lmaydi.
shuffle=True: Bo'laklarga ajratishdan oldin hamma ma'lumotlarni kartaday qorishtirib tashlash. Masalan, tepada faqat yoshlar, pastroqda esa qariyalar tizilib qolmagan bo'lishi ehtimolini oldini olayapti. Turli qatorlar aralashtirilib tarqatiladi.
random_state=42: Aralashtirilgan tartib har safar programmangiz ishlaganida bir xilda ushlab qolinishi (eslab qolinishi). Raqam (taxminan 42 yozilishi an'ana) qanday bo'lishidan qat'iy nazar, aniq natija beradi.
2️⃣ cross_val_score moduli nima qiladi?
cross_val_score funksiyasi yuqorida aytilgan hamma og'ir jarayonlarni kod yozmasdan, bir qator kodda avtomatlashtirishga imkon yaratadigan sehrli instrument.

Siz darsdagi logistik regressiyada shunday yozgan edingiz: scores = cross_val_score(log_reg, x, y, cv=kf, scoring='accuracy')

Talabalarga ushbu argumentlarni ichiga qanday ma'lumot berilayotganini tushuntiramiz:

log_reg – O'rganishi kerak bo'lgan jonsiz model (siz uni LogisticRegression() dan chaqirgan edingiz). "Menga mana shu modelni baholab bering!"
x va y – O'qiladigan barcha savollar (x features) va ularni javoblari (y target).
cv=kf – CV (Cross Validation) ya'ni Nechta bo'lakka bo'lish qoidasi beriladi (tepadagi tayyorlagan kf shartimiz bu yerga kiritiladi). "Mana shu usul asosida uni 5-marotaba imtihon qilgin".
scoring='accuracy' – Baholash mezoni. "Metrika nima orqali bilinsin?". Model Logistik klassifikatsiya bo'lgani uchun aniqlik (Accuracy) olingan (agar Linear Regressionda bo'lganda Xatolik - error beriladi).
Natija nima chiqadi? cross_val_score ishga tushirilsa, bitta javob emas, butun boshli ro'yxat (array) qaytaradi! Bizning holatda n_splits=5 edi, demak u 5 xil urunish baholarini ro'yxat ko'rinishida beradi. Masalan: [0.992, 0.9885, 0.989, 0.984, 0.982].

Talaba undan so'raydi: Xo'sh, modelimiz necha foiz to'g'ri toparkan? Natija 5 ta-ku!? Shunda biz ro'yhat o'rtachasini olamiz - o'rtacha arifmetik natija scores.mean() (barcha baholarni qo'shib, soniga bo'lish). Natijada biz model haqiqiy o'rtacha aniqlikda ~98% aniqlikga ega deymiz va ana endi bu haqiqiy va ishonchli xulosa ekanligiga hech qanday shubhamiz bo'lmaydi!
