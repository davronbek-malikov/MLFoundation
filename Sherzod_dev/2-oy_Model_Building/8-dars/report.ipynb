##  1. Missing Value nima?

###  Ta’rif
**Missing value** — bu datasetdagi mavjud bo‘lishi kerak bo‘lgan, lekin yo‘q (tushib qolgan) qiymatlar.

###  Aniqlash usuli
Missing valuelar odatda maxsus funksiyalar orqali aniqlanadi (masalan, null qiymatlarni tekshirish orqali).

###  To‘ldirish usullari
- **Mean (o‘rtacha qiymat)** — sonli ma’lumotlar uchun ishlatiladi  
- **Median** — outlier (chetki qiymatlar) bo‘lsa samaraliroq  
- **Mode (eng ko‘p uchraydigan qiymat)** — kategorik ma’lumotlar uchun  
- **Forward / Backward fill** — vaqt ketma-ketligi (time-series) uchun  

###  Nima uchun kerak?
- Model noto‘g‘ri ishlamasligi uchun  
- Data yo‘qotilishini oldini olish uchun  

---

## 2. Train-Test Split vs K-Fold Cross Validation

###  Train-Test Split

####  Ishlash prinsipi
Dataset 2 qismga bo‘linadi:
- Training set (model o‘rganadi)
- Testing set (model tekshiriladi)

Odatda 70/30 yoki 80/20 nisbatda bo‘linadi.

####  Xususiyatlari
- Oddiy va tez ishlaydi  
- Natija bitta bo‘linishga bog‘liq  
- Bias (xatolik) yuqori bo‘lishi mumkin  

---

###  K-Fold Cross Validation

####  Ishlash prinsipi
- Dataset K ta teng qismga bo‘linadi  
- Har safar 1 qism test, qolganlari train bo‘ladi  
- Jarayon K marta takrorlanadi  

####  Xususiyatlari
- Har bir data train va test sifatida ishlatiladi  
- Natija aniqroq va ishonchli  
- Hisoblash ko‘proq vaqt oladi  

---

### ⚖️ Asosiy farqlar

| Xususiyat        | Train-Test Split | K-Fold Cross Validation |
|-----------------|----------------|------------------------|
| Tezlik          | Tez            | Sekinroq               |
| Aniqlik         | Pastroq        | Yuqoriroq              |
| Data ishlatilishi | Qisman        | To‘liq                 |
| Ishonchlilik    | Pastroq        | Yuqoriroq              |

##  3. Scaling (Masshtablash) usullari

### Maqsad
Feature (xususiyat) qiymatlarini bir xil diapazonga keltirish orqali modelni yaxshiroq ishlashini ta’minlash.

---

###  StandardScaler

#### G‘oya
- Ma’lumotlar o‘rtacha qiymati 0 ga tenglanadi  
- Standart og‘ish (std) 1 ga teng bo‘ladi  

#### ⚙️ Qachon ishlatiladi?
- Logistic Regression  
- SVM (Support Vector Machine)  

---

### MinMaxScaler

####  G‘oya
- Barcha qiymatlar 0 va 1 oralig‘iga siqiladi  

####  Qachon ishlatiladi?
- KNN (K-Nearest Neighbors)  
- Neural Networks  

---

### 🔹 RobustScaler

####  G‘oya
- Median va IQR (interquartile range) asosida ishlaydi  
- Outlierlarga chidamli  

#### ⚙️ Qachon ishlatiladi?
- Datasetda outlierlar ko‘p bo‘lsa
