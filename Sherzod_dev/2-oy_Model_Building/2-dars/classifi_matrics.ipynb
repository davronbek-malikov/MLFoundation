```python
# Tayyor AI kutubxonasidan asboblarni chaqirib olamiz
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

# Bizning ma'lumotlarimiz
y_actual =  [0, 1, 1, 1, 1, 1, 0, 1, 1, 1]
y_predict = [0, 0, 1, 1, 0, 0, 1, 1, 1, 0]

# Shunchaki tayyor funksiyalarga ma'lumotlarimizni beramiz
tn, fp, fn, tp = confusion_matrix(y_actual, y_predict).ravel()

print(f"TP={tp}, TN={tn}, FP={fp}, FN={fn}\n")

print(f"Accuracy: {accuracy_score(y_actual, y_predict)}")
print(f"Precision: {precision_score(y_actual, y_predict)}")
print(f"Recall: {recall_score(y_actual, y_predict)}")
print(f"F1-Score: {f1_score(y_actual, y_predict):.3f}")
```
