# Data Processing va Preprocessing

## Data Processing (Ma’lumotlarni tayyorlash)
Data processing — bu ma’lumotlarni tahlil yoki mashinaviy o‘rganishdan oldin  
tozalash va foydalanishga tayyorlash jarayoni.

### Asosiy ishlar
- Missing value (tushurib qoldirilgan qiymatlar) bilan ishlash
- Statistik qiymatlar orqali to‘ldirish
- Keraksiz ustun yoki qatorlarni o‘chirish
- Qo‘lda (manual) tuzatish

---

# Missing Values (Tushurib qoldirilgan qiymatlar)

Ma’lumot ichida bo‘sh (NaN) qiymatlar bo‘lsa,  
ularni to‘ldirish yoki o‘chirish kerak bo‘ladi.

## 1) Statistik qiymatlar orqali to‘ldirish

### Mean (o‘rtacha qiymat)
- Faqat sonli ustunlarda ishlaydi
- String (matn) ustunlarda ishlamaydi
- O‘rtacha qiymat bilan to‘ldiradi

"df['col'] = df['col'].fillna(df['col'].mean())"

---

### Median (o‘rtadagi qiymat)
- O‘rtada turgan qiymat bilan to‘ldiradi
- Outlier (juda katta yoki kichik qiymatlar) bo‘lsa eng yaxshi usul
- Sonli ustunlarda ishlaydi

"df['col'] = df['col'].fillna(df['col'].median())"

---

### Mode (eng ko‘p uchraydigan qiymat)
- Eng ko‘p takrorlangan qiymat bilan to‘ldiradi
- String va kategoriyali ustunlar uchun yaxshi

"df['col'] = df['col'].fillna(df['col'].mode()[0])"

---

## 2) Describe (statistik ma’lumotlarni ko‘rish)

Ma’lumotlar haqida umumiy statistikani ko‘rsatadi:
- min : eng kichik qiymat
- max : eng katta qiymat
- 0.25 : 25%
- 0.5 : median
- 0.75 : 75%

"df.describe()"

---

## 3) Fixed value (qo‘lda to‘ldirish)

Ba’zi joylarda qiymatni qo‘lda beramiz  
Masalan: bo‘sh joylarni "No" bilan to‘ldirish

"df['col'].fillna('No', inplace=True)"

---

## 4) Drop (o‘chirish)

### Qatorlarni o‘chirish
Agar qiymatlar juda ko‘p yo‘q bo‘lsa, qatorni o‘chirib yuborish mumkin

"df.dropna(inplace=True)"

Ehtiyot bo‘lish kerak, chunki ma’lumot kamayib ketishi mumkin

---

### Ustunni o‘chirish
Keraksiz ustunni olib tashlash

"df.drop('column_name', axis=1, inplace=True)"

axis=1 → ustun o‘chadi  
axis=0 → qator o‘chadi

---

# Data Preprocessing asosiy jarayoni

1. Ma’lumotni ko‘rish  
   → describe(), info()

2. Missing value tekshirish  
   → isnull(), sum()

3. Missing value to‘ldirish  
   → mean / median / mode / fillna

4. Keraksiz ma’lumotni o‘chirish  
   → drop()

5. Ma’lumot tayyor → modelga berish mumkin
