```python
import pandas as pd

df = pd.read_csv("./../Data/bundesliga_player.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop(["full_name","contract_expires","outfitter","player_agent"], axis=1, inplace=True)
```

```python
df.isnull().sum()
```

```python
df["place_of_birth"].fillna(df["place_of_birth"].mode()[0], inplace=True)
df["foot"].fillna(df["foot"].mode()[0], inplace=True)
df["price"].fillna(df["price"].median(), inplace=True)
df["max_price"].fillna(df["max_price"].median(), inplace=True)
df.isnull().sum()
```
