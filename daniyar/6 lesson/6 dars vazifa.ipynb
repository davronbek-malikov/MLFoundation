1. Mean orqali to’ldiring

```python
import pandas as pd
```

```python
df=pd.read_csv('Sales_with_NaNs_v1.3.csv')
```

```python
df.info()
```

```python
df.nunique()
```

3. Median orqali to’ldiring

```python
df['Sales_After'].fillna(df['Sales_After'].median(),inplace=True)
```

2. Mode orqali to’ldiring

```python
df['Purchase_Made'].fillna(df['Purchase_Made'].mode()[0],inplace=True)
```

```python
df.info()
```

4. Fixed qiymat bn to’ldiring

```python
df['Group'].fillna('Control',inplace=True)
```
