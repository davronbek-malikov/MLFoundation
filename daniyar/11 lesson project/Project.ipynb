```python
df.drop(columns=['index','dateCrawled','abtest','dateCreated','nrOfPictures','lastSeen','name','offerType'],inplace=True)
```

```python
for col in df.columns:
    if df[col].dtype=='object':
        df[col].fillna(df[col].mode()[0],inplace=True)
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder=LabelEncoder()
```

```python
df[col] = encoder.fit_transform(df[col])     # kere emas 
```

```python
for col in df.columns:
    if df[col].dtype=='object':
        if df[col].nunique()<=10:
            dummies=pd.get_dummies(df[col],prefix=col,dtype=int)
            df=pd.concat([df.drop(columns=col),dummies],axis=1)
       
        else:
            df[col]=encoder.fit_transform(df[col])
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
num_col=df.select_dtypes(include=['int64','float64']).columns.drop(['price'])
df[num_col]=scaler.fit_transform(df[num_col])
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
x=df.drop('price',axis=1)
y=df['price']
```

```python
from sklearn.model_selection  import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
y_pred
```

```python

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
mae=mean_absolute_error(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
```

```python
print('mae:',mae)
print('mse:',mse)
print('r2:',r2)
```
