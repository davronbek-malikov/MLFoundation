# Data Preprocessing 

```python

df.drop(['index'],axis=1,inplace=True)
df.drop(['dateCrawled'],axis=1,inplace=True)
df.drop(['lastSeen'],axis=1,inplace=True)
df.drop(['name'],axis=1,inplace=True)
df.drop(['seller'],axis=1,inplace=True)
df.drop(['offerType'],axis=1,inplace=True)
df.drop(['postalCode'],axis=1,inplace=True)
df.drop(['nrOfPictures'],axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.nunique(
    
)
```

```python
for col in df.columns:
    if df[col].isnull().sum():
         if df[col].dtype == 'str':
              df[col]=df[col].fillna(df[col].mode()[0])

   
```

```python
df.isnull().sum()
```

# Encoding 

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()

for col in df.columns:
    if df[col].dtype=='str':
        df[col]=encoder.fit_transform(df[col])
```

```python
df.info()
```

# Scaling

```python
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
num_col=df.select_dtypes(include=['int64']).columns.drop('price')
for col in num_col:
    df[num_col]=scaler.fit_transform(df[num_col])   
```

```python
df.head()
```

# Model selection

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr
```

```python
x=df.drop('price',axis=1)
```

```python
y=df['price'].astype(int)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
df.info()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

# Evaluation

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
mse=mean_squared_error(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
```

```python
print('mae:',mae)
print('mse:',mse)
print('r2:',r2)
```
