```python
import pandas as pd
```

```python
df=pd.read_csv("oraliq imtihon 6-dars/netflix_titles.csv")
```

```python
df.info()
```

```python
df.isnull().sum()
```

1. Mean orqali to’ldiring

```python
df["release_year"]=df["release_year"].fillna(df["release_year"].mean())
```

```python
df.isnull().sum()
```

2. Mode orqali to’ldiring

```python
df['director']=df['director'].fillna(df['director'].mode()[0])
```

```python
df.isnull().sum()
```

3. Median orqali to’ldiring

ozgina tuzatish kiritamiz bu yerda bittagina int typedagi ustun bo'lganligi uchun yana 2tasini intga o'tkazishimiz kerak chunki bizda median va mean faqat numerical qiymatlar bilan ishlay oladi xolos

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
df['rating'] = encoder.fit_transform(df['rating'])
```

```python
df.info()
```

```python
df['rating']=df['rating'].fillna(df['rating'].median())
```

```python
df.isnull().sum()
```
