```python
import numpy as np

y_actual = np.array([2, 5, 7, 3, 9.4])
y_pred   = np.array([3, 4, 6, 2, 8.7])
```

```python
mse = np.mean((y_actual - y_pred) ** 2)
mse
```

```python
rmse = np.sqrt(mse)
rmse
```

```python
mae = np.mean(np.abs(y_actual - y_pred))
mae
```

```python
r2 = 1 - (np.sum((y_actual - y_pred) ** 2) / np.sum((y_actual - np.mean(y_actual)) ** 2))
r2
```
