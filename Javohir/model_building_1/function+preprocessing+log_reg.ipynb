# Missing value + function

```python
def toldir(df):
    for col in df.columns:
        if df[col].isnull().any():
            if df[col].dtype== 'str':
                df[col]=df[col].fillna(df[col].mode()[0])
            else:
                df[col]=df[col].fillna(df[col].mean())
    return df       
```

```python
toldir(df)
```

```python
df.isnull().sum()
```

# Encoding

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
```

# Encoding + Function

```python
def raqamliga(df):
    for col in df.columns:
        if df[col].dtype=='str':
            df[col]=encoder.fit_transform(df[col])
    return(df)
```

```python
raqamliga(df)
```

# Scaling

```python
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler
scaler=MinMaxScaler()
```

# Scalimg + function

```python
def scaling_qil(df):
    num_col=df.select_dtypes(include=['int64','float64']).columns.drop('cgpa')
    for col in num_col:
        df[col]=scaler.fit_transform(df[[col]])
    return df
```

```python
scaling_qil(df)
```

# Model Training

```python
x=df.drop('burnout_level',axis=1)
```
