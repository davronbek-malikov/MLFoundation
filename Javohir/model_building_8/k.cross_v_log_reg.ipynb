```python
import pandas as pd
```

```python
df=pd.read_csv('amazon-fashion-800k+-user-reviews-dataset.csv')
# Amazon Fashion 800k+ User Reviews Dataset
# Large Fashion Review Dataset for Sentiment Analysis and Rating Prediction
```

```python
df.info()
```

# Data Preprocessing 

```python
df.isnull().sum()
```

```python
df.drop('title', axis=1, inplace=True) # bu predict qilishimiz uchun keraksiz column
```

```python
# Missing 
df['text']=df['text'].fillna(df['text'].mode()[0])
```

```python
df.isnull().sum()
```

```python
# Encoding 
from sklearn.preprocessing import LabelEncoder
df.info()
```

```python
def encodla(df):
    encoder = LabelEncoder()

    for col in df.columns:

        if df[col].dtype == 'bool':
            df[col] = df[col].astype(int)

        elif df[col].dtype == 'str':

            if df[col].nunique() <= 5:
                dummies = pd.get_dummies(df[col], prefix=col)
                df = pd.concat([df.drop(columns=[col]), dummies], axis=1)

            else:
                df[col] = encoder.fit_transform(df[col])

    return df
```

```python
df=encodla(df)
```

```python
df.head()
```

```python
df.info()
```

```python
# scaling 
from sklearn.preprocessing import MinMaxScaler

def scale_qil(df):
    scaler= MinMaxScaler()
    num_col=df.select_dtypes(include=['int64','float64']).columns.drop('target')
    df[num_col]=scaler.fit_transform(df[num_col])
    return df
```

```python
df=scale_qil(df)
```

```python
df.head()
```

# Model


```python
x=df.drop('target', axis=1)
```

```python
y=df['target']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# linear regression
from sklearn.linear_model import LogisticRegression
log_reg=LogisticRegression()
```

# Logistic Regrassion

```python
log_reg
```

```python
log_reg.fit(x_train,y_train)
```

```python
y_pred=log_reg.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
score=accuracy_score(y_test,y_pred)
class_report=classification_report(y_test,y_pred)
```

```python
print('Modelning aniqlik darajasi:',score)
```

# k_Fold validation

```python
from sklearn.model_selection import  KFold, cross_val_score
kf=KFold(n_splits=5,shuffle=True,random_state=42)
```

```python
kf
```

```python
scores=cross_val_score(log_reg,x,y,cv=kf,scoring='accuracy')
print(scores)
```
