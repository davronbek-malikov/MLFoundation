# Encoding 

```python
import pandas as pd 
df=pd.read_csv('imdb_top_1000.csv')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Gross'] = df['Gross'].fillna(df['Gross'].mode()[0]) 
```

```python
df.info()
```

```python
df['Meta_score'] = df['Meta_score'].fillna(df['Meta_score'].mean())
```

```python
df.info()
```

```python
df.drop('Certificate',axis=1,inplace=True)
```

```python
df.info()
```

```python
df.nunique()
```

# One-hot encoding

```python
dummies=pd.get_dummies(df['IMDB_Rating'],prefix='ustun',dtype=int)
```

```python
dummies
```

```python
df=pd.concat([df.drop(columns=['IMDB_Rating']),dummies],axis=1)
```

```python
df.info()
```
