```python
# def + for + if
def juft_sonlarni_chiqar(sonlar):
    for son in sonlar:     
        if son % 2 == 0:    
            print(son, "juft")
        else:            
            print(son, "toq")

sonlar = [77,45,8,65,99]

juft_sonlarni_chiqar(sonlar)
```
