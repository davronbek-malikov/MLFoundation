# Funksiyaga oid misollar 

```python
# 2 ta sozni jamlovchi funksiya 
def jamla(x,y):
    return x+''+y
```

```python
jamla('Sa','lom')
```

```python
# sonni qoshadigan funksiya 
def jamla(a,b,c):
    return a+b+c
```

```python
jamla(1,2,3)
```

```python
# sonni ayradigan funksiya 
def ayr(a,b,c,d):
    return a-b-c-d 
```

```python
ayr(10,2,3,1)
```

```python
# def + for 
def bolish(a, s):
    return a / s

sonlar = [20, 10] 

for son in sonlar:
    print(bolish(son,4))
```

```python
# def + if 
def juft_toq(x):
    if x % 2 == 0:
        return "juft"
    else:
        return "toq"

sonlar = [1,4,12,3,6]

for son in sonlar:
    print(son, juft_toq(son))
```

```python
# def + for + if
def juft_sonlarni_chiqar(sonlar):
    for son in sonlar:     
        if son % 2 == 0:    
            print(son, "juft")
        else:            
            print(son, "toq")

sonlar = [11,23,1,44,12,4]

juft_sonlarni_chiqar(sonlar)
```
