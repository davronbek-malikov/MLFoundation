# Data bilan tanishuv 

```python
import pandas as pd 
df=pd.read_csv('user_music_listen_data.csv')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.drop("User_ID", axis=1)
```

```python
def toldir(df):
    for col in df.columns:
        if df[col].isnull().any():
            if df[col].dtype == 'object':
                df[col] = df[col].fillna(df[col].mode()[0])
            else:
                df[col] = df[col].fillna(df[col].mean())
    return df            
```

```python
toldir(df)
```
