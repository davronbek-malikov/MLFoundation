# Data preprocessing + For loop

```python
import pandas as pd 
df=pd.read_csv('user_music_listen_data.csv')
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['Imag_Dragons'] = df['Imag_Dragons'].fillna(df['Imag_Dragons'].mode())
```

```python
df['T_Scott'] = df['T_Scott'].fillna(df['T_Scott'].median())
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
for col in df.columns:
    if df[col].dtype == 'float64':
        df[col] = df[col].fillna(df[col].median())
```

```python
df.isnull().sum()
```
