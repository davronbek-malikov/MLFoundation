```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
from sklearn.impute import SimpleImputer

# Kaggle'dan Titanic dataset
df = pd.read_csv("titanic.csv")

# Missing values tozalash
strategies = {
    "mean": SimpleImputer(strategy="mean"),
    "median": SimpleImputer(strategy="median"),
    "mode": SimpleImputer(strategy="most_frequent")
}

for col in df.select_dtypes(include=['float64', 'int64']).columns:
    for name, imputer in strategies.items():
        df[col] = imputer.fit_transform(df[[col]])

# Encoding
for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].astype('category').cat.codes

# Scaling
scalers = {
    "standard": StandardScaler(),
    "minmax": MinMaxScaler(),
    "robust": RobustScaler()
}

scaled_dfs = {}
for name, scaler in scalers.items():
    scaled_dfs[name] = pd.DataFrame(
        scaler.fit_transform(df.select_dtypes(include=['float64', 'int64'])),
        columns=df.select_dtypes(include=['float64', 'int64']).columns
    )
```

# Scaling (3 usul)

Natija:

StandardScaler: qiymatlar o‘rtacha 0, dispersiya 1 bo‘ldi.

MinMaxScaler: qiymatlar 0–1 oralig‘iga tushdi.

RobustScaler: median asosida masshtablash qilindi, outlierlarga chidamli.

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler

scalers = {
    "standard": StandardScaler(),
    "minmax": MinMaxScaler(),
    "robust": RobustScaler()
}

scaled_dfs = {}
for name, scaler in scalers.items():
    scaled_dfs[name] = pd.DataFrame(
        scaler.fit_transform(df.select_dtypes(include=['float64', 'int64'])),
        columns=df.select_dtypes(include=['float64', 'int64']).columns
    )
    print(f"\n{name.upper()} Scaling natijasi:\n", scaled_dfs[name].head())
```
