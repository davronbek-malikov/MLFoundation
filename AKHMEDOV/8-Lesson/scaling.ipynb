```python
for col in df.columns:
    if df[col].isnull().any():
        if df[col].dtype=='object': 
            df[col].fillna(df[col].mode()[0], inplace=True)
        else:
            df[col].fillna(df[col].mean(), inplace=True)
```

# for loop + scaling

```python
scaler = MinMaxScaler()
for col in df.columns:
    if df[col].dtype != 'object':
        df[col] = scaler.fit_transform(df[[col]])
```
