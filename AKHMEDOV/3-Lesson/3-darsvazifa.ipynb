```python
# Cell 2: Pasport olish sharti (safer input handling)
# If the user presses Enter, default age 30 will be used. Invalid input is handled.
try:
    s = input('Yoshingizni kiriting (default 30): ').strip()
    yosh = int(s) if s != '' else 30
except ValueError:
    print("Noto'g'ri kiritish — yosh butun son bo'lishi kerak. Default 30 ishlatiladi.")
    yosh = 30


hujjat = input("Kerakli hujjatlar topshirildimi? (ha/yo'q): ").strip()
rasm = input("3x4 rasm topshirildimi? (ha/yo'q): ").strip()

if yosh >= 16 and hujjat.lower() == "ha" and rasm.lower() == "ha":
    print("Siz pasport olishingiz mumkin.")
elif yosh < 16:
    print("Yoshingiz pasport olish uchun yetarli emas.")
else:
    print("Kerakli hujjatlar to'liq emas.")
```

```python
# Cell 1: Imtihondan o'tishni nazorat qilish (safer input handling)
# Accepts empty input (uses defaults) and handles non-integer input without crashing.
try:
    s = input("Ballni kiriting (30, 60 yoki 90) [default 60]: ") .strip()
    ball = int(s) if s != '' else 60
except ValueError:
    print("Noto'g'ri kiritish — ball butun son bo'lishi kerak. Default 60 ishlatiladi.")
    ball = 60

try:
    t = input("Darsga qatnashish foizini kiriting (0-100) [default 100]: ") .strip()
    qatnashish = int(t) if t != '' else 100
except ValueError:
    print("Noto'g'ri kiritish — foiz butun son bo'lishi kerak. Default 100 ishlatiladi.")
    qatnashish = 100

# Ensure values are in sensible ranges (optional)
if qatnashish < 0:
    qatnashish = 0
if qatnashish > 100:
    qatnashish = 100

if ball >= 60 and qatnashish >= 70:
    print("Tabriklaymiz! Siz imtihondan o'tdingiz.")
elif ball >= 60 and qatnashish < 70:
    print("Siz baldan o'tdingiz, lekin qatnashish foizi yetarli emas.")
else:
    print("Afsuski, siz imtihondan o'ta olmadingiz.")
```
