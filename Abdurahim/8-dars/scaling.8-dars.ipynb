```python
import pandas as pd
```

```python
df=pd.read_csv('earthquake_data_tsunami.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().any()
```

```python
df.isnull().sum()
```

```python
df.drop('Month', axis=1, inplace =True)
```

```python
df['Year'].fillna(df['Year'].mode()[0], inplace=True)
```

```python
df.columns
```

```python
df['gap'].fillna((30), inplace=True)
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
for col in df.columns:
    if df[col].isnull().any:
        if df[col].dtype=='int' and 'float64':
            df[col].fillna(df[col]. mean(),  inplace=True)
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
```

```python
num_col=df.select_dtypes(include=['int64', 'float64']).columns.drop('tsunami')
```

```python
num_col
```

#StandardScaler

```python
scaler=StandardScaler()
```

```python
scaler
```

```python
df['Year']=scaler.fit_transform(df[['Year']])
```

```python
df.head()
```

MinMaxScaler
