```python
for col in df.columns:
    if df[col].isnull().sum() > 0:
        if df[col].dtype=='object':
            df[col].fillna(df[col].mode()[0],inplace=True)
        else:df[col].fillna(df[col].mean(),inplace=True)
    else:df[col].fillna(df[col].median(),inplace=True)
```
