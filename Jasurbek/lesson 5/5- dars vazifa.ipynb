```python
import pandas as pd
# df ni ozi ishlamadi printni olib keldim.
df = pd.read_csv('apple_5yr_one.csv') 
print(df.isnull().sum())
```

```python
columns = ["Open"]
df.drop(columns=columns, inplace=True)
```

**# One- Hot encoding**

```python
df.columns = df.columns.str.strip()
```

```python
print(df.columns)
```

```python
# data harfini kichik bilan yoganim uchun xatolik bo'lgan ekan
dummies = pd.get_dummies(df["Date"], prefix="Low", dtype=int)
dummies
```

```python
df = pd.concat([df, dummies], axis=1)
```
