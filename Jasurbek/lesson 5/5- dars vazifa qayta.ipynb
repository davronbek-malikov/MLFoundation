```python
import pandas as pd
# df ni ozi ishlamadi printni olib keldim.
df = pd.read_csv('apple_5yr_one.csv') 
print(df.isnull().sum())
```

```python
df.head()
```

Drop columns

```python
columns = ["Open"]
df.drop(columns=columns, inplace=True)
```

```python
print(df.isnull().sum())
```

```python
df.info()
```

```python
df.head()
```

```python
df.nunique()
```

Mean

```python
df["Date"] = pd.to_numeric(df["Date"], errors='coerce')
```

```python
# 1255 ortacha qiymat lekin buni ortca qiymay deb hisoblamadi
df["Date"].fillna(df["Date"].mean(), inplace=True)
```

Mode

```python
#low
df["Low"].fillna(df["Low"].mode()[0], inplace=True)
df["High"].fillna(df["High"].mode()[0], inplace=True)
```

Median

```python
# buni ham songa otkazdim
df["High"] = pd.to_numeric(df["High"], errors='coerce')
```

```python
df["High"].fillna(df["High"].median(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.nunique()
```

One- Hot encoding

```python
df.columns = df.columns.str.strip()
```

```python
print(df.columns)
```

```python
# data harfini kichik bilan yoganim uchun xatolik bo'lgan ekan
dummies = pd.get_dummies(df["Date"], prefix="Low", dtype=int)
dummies
```

```python
dummies = pd.get_dummies(df["Date"], prefix="Date")
```

```python
df = pd.concat([df, dummies], axis=1)
```

```python
df=pd.concat([df.drop(columns=["Date"]), dummies], axis=1)
```

```python
df.info()
```

```python
df.nunique()
```

Label encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
```

```python
df["High"] = encoder.fit_transform(df["High"])
df["Low"] = encoder.fit_transform(df["Low"])
df["Close"] = encoder.fit_transform(df["Close"])
df[ "Volume"] = encoder.fit_transform(df["Volume"])
```
