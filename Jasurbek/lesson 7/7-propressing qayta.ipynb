```python
import pandas as pd
```

```python
df = pd.read_csv("apple_5yr_one.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.dtypes
```

```python
columns=df.columns

for col in columns:
	if df[col].isnull().any():
		if df[col].dtype == "float":
			if col == 'Date':
				df[col].fillna(df[col].median(), inplace=True)
			else: df[col].fillna(df[col].mean(), inplace=True)
		else: df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
dummies = pd.get_dummies(df["Date"], prefix="col", dtype=int)
```

```python
df=pd.concat([df.drop(columns=["Date"]), dummies], axis=1)
```
