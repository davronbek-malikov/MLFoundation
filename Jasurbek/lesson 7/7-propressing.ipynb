```python
import pandas as pd
```

```python
df = pd.read_csv("Player.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.dtypes
```

```python
columns=df.columns

for col in columns:
	if df[col].isnull().any():
		if df[col].dtype == "float":
			if col == 'Unnamed':
				df[col].fillna(df[col].median(), inplace=True)
			else: df[col].fillna(df[col].mean(), inplace=True)
		else: df[col].fillna(df[col].mode()[0], inplace=True)
```
