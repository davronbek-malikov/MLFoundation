# Data yuklash

```python
import pandas as pd
df=pd.read_csv("Social_media_impact_on_life.csv")
```

```python
df.info()
```

# Data preprocessing

```python
df.isnull().sum()
```

```python
# Missing
def toldir(df):
    for col in df.columns:
        if df[col].isnull().any():
            if df[col].dtype=="str":
                df[col]=df[col].fillna(df[col].mode()[0])
            else:
                df[col]=df[col].fillna(df[col].mean())
    return df
```

```python
df=toldir(df)
```

```python
df.isna().any()
```

```python
# Encoding
df.nunique()
```

```python
from sklearn.preprocessing import LabelEncoder
def encodla(df):
    encoder=LabelEncoder()
    for col in df.columns:
        if df[col].dtype=="str":
            if df[col].nunique()<=5:
                dummies=pd.get_dummies(df[col],prefix=col,dtype=int)
                df=pd.concat([df.drop(columns=[col]),dummies],axis=1)
            else:
                df[col]=encoder.fit_transform(df[col])
    return df
```

```python
df=encodla(df)
```

```python
df.info()
```

```python
df.head()
```

```python
df.tail()
```

```python
# Scaling
from sklearn.preprocessing import MinMaxScaler
def scalqil(df):
    scaler=MinMaxScaler()
    num_col=df.select_dtypes(include=['int64','float64']).columns.drop("Affects_Academic_Performance_Yes")
    df[num_col]=scaler.fit_transform(df[num_col])
    return df
```

```python
df=scalqil(df)
```

```python
df.head()
```

# Training

```python
x=df.drop(columns=['Affects_Academic_Performance_Yes','Affects_Academic_Performance_No'])
y=df['Affects_Academic_Performance_Yes']
```

# Eski usulda o`zimiz topish orqali

```python
from sklearn.model_selection import train_test_split
```

```python
# x_train,x_test=x
# y_train,y_test=y
x_train,x_test,y_train,y_test=train_test_split(
    x,y,test_size=0.2,random_state=42
    )
```

```python
# Model(algorithm chaqirish)
from sklearn.linear_model import LogisticRegression
log_reg=LogisticRegression()
```

```python
log_reg
```

```python
log_reg.fit(x_train,y_train)
```

```python
# Predict qilish (modelning natijasini taxmin qilish)
y_pred=log_reg.predict(x_test)
```

```python
# Metricslar orqali baholash (Evaluation)
from sklearn.metrics import accuracy_score
```

```python
accuracy=accuracy_score(y_test,y_pred)
```

```python
print(accuracy)
```

```python
# K-Fold cross validation
from sklearn.model_selection import KFold, cross_val_score
```

```python
kf=KFold(n_splits=5,shuffle=True,random_state=42)
```

```python
kf
```

```python
scores=cross_val_score(log_reg,x,y,cv=kf,scoring='accuracy')
```
