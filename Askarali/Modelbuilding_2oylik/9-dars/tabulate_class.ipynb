# Classification dataset yuklab (Tabulate)

```python
import pandas as pd
df=pd.read_csv("Social_media_impact_on_life.csv")
```

```python
df.info()
```

```python
df.columns
```

```python
df.drop('Student_ID',inplace=True,axis=1)
```

```python
df.isnull().sum()
```

```python
# Missinfg value
def toldir(df):
    for col in df.columns:
        if df[col].isnull().any():
            if df[col].dtype=='str':
                df[col]=df[col].fillna(df[col].mode()[0])
            else:
                df[col]=df[col].fillna(df[col].mean())
    return df
```

```python
df=toldir(df)
```

```python
df.isna().any()
```

```python
df.nunique()
```

```python
# Encoding
from sklearn.preprocessing import LabelEncoder
def encodla(df):
    encoder=LabelEncoder()
    for col in df.columns:
        if df[col].dtype=="str":
            if df[col].nunique()<=3:
                dummies=pd.get_dummies(df[col],prefix=col,dtype=int)
                df=pd.concat([df.drop(columns=[col]),dummies],axis=1)
            else:
                df[col]=encoder.fit_transform(df[col])
    return df
```

```python
df=encodla(df)
```

```python
df.info()
```

```python
df.drop(['Affects_Academic_Performance_No','Gender_Female'],axis=1,inplace=True)
```

```python
df.info()
```

```python
df.head()
```

```python
# Scaling
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler
def scal_qil(df):
    scaler=MinMaxScaler()
    num_col=df.select_dtypes(include=['int64','float64']).columns.drop('Affects_Academic_Performance_Yes')
    df[num_col]=scaler.fit_transform(df[num_col])
    return df
```

```python
df=scal_qil(df)
```

```python
# Input va Output qiymatlarga ajratin olamiz
x=df.drop('Affects_Academic_Performance_Yes',axis=1)
y=df['Affects_Academic_Performance_Yes']
```

```python
# Model_selection ya'ni train, test va splitga ajratib olish
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

# Logistic Regression

```python
# Model(algorithm) chaqirish
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
lr
```

```python
# Modelni o'qitish
lr.fit(x_train,y_train)
```

```python
# Predict qilish
y_pred=lr.predict(x_test)
```

```python
# Evaluation yani metrics orqali tekshirish
from sklearn.metrics import classification_report, confusion_matrix
```

```python
log_cr=classification_report(y_test,y_pred)
log_cm=confusion_matrix(y_test,y_pred)
```
