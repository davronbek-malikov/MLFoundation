# Function

```python
# 1. Son musbat yoki manfiy
def check_sign(number):
    return 'positive' if number > 0 else 'negative'

print(check_sign(-5))
```

```python
# 2. Son 10 dan katta yoki kichik
def kichik_10_katta(x):
    return 'big' if x > 10 else 'small'

print(kichik_10_katta(11))
```

```python
# 3. Son 0 ga tengmi
def is_zero(n):
    return 'zero' if n == 0 else 'not zero'

print(is_zero(0))
```

```python
# 4. Ikki sonning kattasini topish
def max_number(a, b):
    return a if a > b else b

print(max_number(3, 8))
```

```python
# 5. Son 5 ga bo‘linadimi
def bolinadi_five(n):
    return 'yes' if n%5==0 else 'no'

print(bolinadi_five(15))
```

```python
# 6. String uzunligini tekshirish
def string_length(s):
    return 'long' if len(s) > 5 else 'short'

print(string_length("hello"))
```

```python
# 7. Son kvadratini topish
def kvadrat(n):
    return n * n

print(kvadrat(4))
```

```python
# 8. Son manfiy bo‘lsa musbatga o‘tkazish
def manfiydan_musbatga(n):
    return -n if n < 0 else n

print(manfiydan_musbatga(-9))
```

```python
# 9. Ikki son tengligi
def equal(a, b):
    return 'equal' if a==b else 'not equal'

print(equal(5, 5))
```

```python
# 10. Son 3 va 5 ga bo‘linadimi
def bolinadi_3_5(n):
    return 'yes' if n%3==0 and n%5==0 else 'no'

print(bolinadi_3_5(15))
```
