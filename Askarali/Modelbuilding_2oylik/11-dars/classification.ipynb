# Classification uchun

```python
import pandas as pd
df=pd.read_csv("digital_mental_health_classification.csv")
```

```python
df.info()
```

```python
df.head()
```

```python
df.tail()
```

```python
# Missing value
df.isna().sum()
```

```python
def fulla(df):
    for col in df.columns:
        if df[col].isnull().any():
            if df[col].dtype=="str":
                df[col]=df[col].fillna(df[col].mode()[0])
            else:
                df[col]=df[col].fillna(df[col].mean())
    return df
df=fulla(df)
```

```python
df.isna().any()
```

```python
df.info()
```

```python
# Encoding jarayoni
from sklearn.preprocessing import LabelEncoder
```

```python
df.nunique()
```

```python
def raqamla(df, Mental_Health_Risk):
    encoder=LabelEncoder()
    for col in df.columns:
        if col == Mental_Health_Risk:
            continue

        if df[col].dtype=="str":
            if df[col].nunique()<=5:
                dummies=pd.get_dummies(df[col],prefix=col,dtype=int)
                df=pd.concat([df.drop(columns=[col]),dummies],axis=1)
            else:
                df[col]=encoder.fit_transform(df[col])
    return df
```

```python
df=raqamla(df, "Mental_Health_Risk")
```

```python
df.info()
```

```python
df.head()
```

```python
# Scaling jarayoni
from sklearn.preprocessing import MinMaxScaler
```

```python
def ritimla(df):
    scaler=MinMaxScaler()
    num_col=df.select_dtypes(include=["float64","int64"]).columns.drop("Mental_Health_Risk")
    df[num_col]=scaler.fit_transform(df[num_col])
    return df
```

```python
df=ritimla(df)
```

```python
df.head()
```

```python
# Target(output) va Feature(input) ga ajratish
from sklearn.model_selection import train_test_split
```

```python
x=df.drop("Mental_Health_Risk",axis=1).astype(int)
y=df["Mental_Health_Risk"].astype(int)
```

```python
x_train,x_test, y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

# `Algorithm(model) tanlash`

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lgr=LogisticRegression()
```

```python
lgr
```

```python
# Fitting(algorithmni o'qitish)
lgr.fit(x_train,y_train)
```

```python
# Predict(baxolash)
y_pred=lgr.predict(x_test)
```

```python
# Evaluation
from sklearn.metrics import classification_report, accuracy_score
```

```python
class_report=classification_report(y_test,y_pred)
acc_score=accuracy_score(y_test,y_pred)
```

```python
print(f'class_report: {class_report}')
print("acc_score: ", acc_score)
```
