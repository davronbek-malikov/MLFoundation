# Preprocessing

```python
df.isnull().any()
```

```python
print(df.dtypes)
```

```python
# missing + for loop
```

```python
for col in df.columns:
    if df[col].isnull().sum():
        if df[col].dtype=="str":
            df[col] = df[col].fillna("Unknown")
            #df[col].fillna("Unknown")
```

```python
#for col in df.columns:
#if df[col].isnull().sum():
#if df[col].dtype=="str":
#df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df.isnull().sum()
```

```python
# encoding
```

```python
df.nunique() #classlar soni
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
encoder
```

```python
# encoding + for loop
for col in df.columns:
    if df[col].dtype=="string":
        if df[col].nunique()<=3:
            dummies=pd.get_dummies(df[col],prefix=col,dtype=int)
            df=pd.concat([df.drop(columns=[col]),dummies],axis=1)
        else:
            df[col] = encoder.fit_transform(df[col])
```

```python
# Model selection
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr
```

```python
# Input(x,feature) va Output(y,target) ajratib olish
```

```python
# Input
x=df.drop("price",axis=1)
```

```python
# Output
y=df["price"]
```

```python
# Train/Test
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
lr.fit(x_train,y_train)
```

```python
# prediction(bashorat qilish)
```

```python
y_predict=lr.predict(x_test)
```

```python
y_predict
```

```python
# Evaluation(bu modelning bashorati qanchalik to‘g‘ri ekanini tekshirish jarayoni)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
mae=mean_absolute_error(y_test,y_predict) #Model o‘rtacha nechchi % xato qilganini ko‘rsatadi
mse=mean_squared_error(y_test,y_predict) #Model qilgan xatolarning o‘rtacha kvadratini topadi(manfiy yoqoladi,katta xatolar ko'proq jazolanadi)
r2=r2_score(y_test,y_predict) # 1ga yaqin → juda yaxshi, 0 → o'rtacha, 0dan kichik → juda yomon
```

```python
print("mae: ",mae)
print("mse: ",mse)
print("r2: ",r2)
```
