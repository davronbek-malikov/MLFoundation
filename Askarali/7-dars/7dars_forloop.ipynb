# `for loop`ga oid misollar

```python
a=[1,2,3]
for i in a:
    print(i)
```

```python
#1 oddiy sonni chiqarish
for i in range(1,6):
    print(i)
```

```python
#2 sonlar yig'indisi
summa = 0

for i in range(1,11):
    summa += i

print("Yig'indi:",summa)
```

```python
#3 faqat juft sonlarni chiqarish
for i in range(1,11):
    if i%2 == 0:
        print(i)
```

```python
#4 ro'yxat elementlarini chiqarish

mevalar = ["olma","banan","uzum"]

for meva in mevalar:
    print(meva)
```

```python
#5 so'z ichidagi harflarni chiqarish

soz = "Python"

for harf in soz:
    print(harf)
```

```python
#6 5karrali jadval chiqarish

for i in range(1,11):
    print("5 x", i, "=", 5 * i)
```

```python
#7 sozning nechta ekanligini topish
mevalar=['olma','banan','uzzum','anor']
for harfsoni in mevalar:
    print(len(harfsoni))
```

```python
#8 faqat toq sonlar
for son in range(1,11):
    if son%2 != 0:
        print(son)
```

```python
#9 7karrali jadval
 
for i in range(1,11):
    print("7 x",i,"=",7*i)
```

```python
#10 2-usulda juft sonlarni topish
for i in range(2,11,2):
    print(i)
```
