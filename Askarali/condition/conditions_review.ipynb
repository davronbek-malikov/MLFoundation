# Condition review

```python
age = int(input("Yoshni kiriting"))
score = int(input("Ballni kiriting"))
attendance = int(input("Davomat foizini kiriting"))
discipline = int(input("Intizom bahosi(1-10): "))

if age >= 18:
    if (score >= 80 and attendance >= 85) or (score >= 90):
        if discipline >= 3:
            result = "Grand asosida o`tdi"
        else:
            result = "Intizom sababli rad etildi"
    elif score >= 60 and attendance >= 75:
        result = "Kontrakt asosida o`tdi"
    else:
        result = "Yiqildi"
else:
    result = "Yosh yetarli emas"

print(result)
```

```python
balance = int(input("Balansni kiriting: "))
withdraw = int(input("Yechmoqchi bo`lgan summa: "))
daily_limit = int(input("Kunlik limit: "))
pin = int(input("Pin kodni kiriting: "))

if pin == 1234:
    if withdraw > 0:
        if withdraw <= daily_limit and withdraw <= balance:
            if (balance - withdraw) >= 50000:
                result = "Pul muvafaqqiyatli yechildi"
            else:
                result = "Minimal balans saqlanishi kerak"
        else:
            result = "Limit yoki balans yetarli emas"
    else:
        result = "Noto`g`ri summa"
else:
    result = "Pin xato"
print(result)
```
