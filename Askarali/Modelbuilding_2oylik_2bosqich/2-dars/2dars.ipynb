```python
# Missing value
def toldir(df):
    for col in df.columns:
        if df[col].isnull().any():
            if df[col].dtype=="str":
                df[col]=df[col].fillna(df[col].mode()[0])
            else:
                df[col]=df[col].fillna(df[col].mean())
    return df
```

```python
df=toldir(df)
```

```python
df.isna().any()
```

```python
df.nunique()
```

```python
df.info()
```

```python
# Encoding
from sklearn.preprocessing import LabelEncoder
def encodla(df):
    # le=LabelEncoder()   ---> # bitta encoderni hamma ustunlar ishlatish uchun
    df = df.copy()
    for col in df.columns:
        if df[col].dtype=="str":
            le=LabelEncoder() # har bir col ustun uchun alohida alohida encoder ishlatish uchun
            df[col]=le.fit_transform(df[col])
    return df
```

```python
df=encodla(df)
```

```python
# Scaling Classification uchun
from sklearn.preprocessing import MinMaxScaler
def scalingqil(df):
    scaler=MinMaxScaler()
    num_col=df.select_dtypes(include=["int64","float64"]).columns.drop("credit_risk")
    df[num_col]=scaler.fit_transform(df[num_col])
    return df
```
