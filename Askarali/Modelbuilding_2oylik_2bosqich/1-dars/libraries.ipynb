```python
import pandas as pd
```

```python
df.drop("User_ID",inplace=True,axis=1)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
# Missing value
def toldir(df):
    for col in df.columns:
        if df[col].isnull().any():
            if df[col].dtype=="str":
                df[col]=df[col].fillna(df[col].mode()[0])
            else:
                df[col]=df[col].fillna(df[col].mean())
    return df
```

```python
df=toldir(df)
```

```python
df.isna().any()
```

```python
df.nunique()
```

```python
df.head()
```

```python
# Encoding
from sklearn.preprocessing import LabelEncoder
def encodla(df):
    encoder=LabelEncoder()
    for col in df.columns:
        if df[col].dtype=="str":
            df[col]=encoder.fit_transform(df[col])
    return df
```

```python
df=encodla(df)
```

```python
df.info()
```

```python
df.head()
```

```python
# Scaling
from sklearn.preprocessing import MinMaxScaler
def scaling_qil(df):
    scaler=MinMaxScaler()
    num_col=df.select_dtypes(include=["int64","float64"]).columns.drop("App_Usage_Count")
    df[num_col]=scaler.fit_transform(df[num_col])
    return df
```

```python
df=scaling_qil(df)
```

```python
# Target va Feature ga ajratish
x = df.drop("App_Usage_Count",axis=1)
y = df["App_Usage_Count"]
```

# Tabulate

```python
from tabulate import tabulate
```

```python
# Classification
results_classification = [
    ['Logistic Regression', 0.50, 0.51, 0.53, 0.52],
    ['Decison Tree Classifier', 0.50, 0.51, 0.50, 0.50],
    ['Random Forest Classifier', 0.50, 0.51, 0.53, 0.52],
    ['Support Vector Classifier', 0.50, 0.51, 0.46, 0.48]
]
```

```python
headers_classification = ['Algorithms', 'Accuracy','Precision','Recall','F1-Score']
table_classification = tabulate(results_classification, headers=headers_classification, tablefmt='grid', floatfmt='.2f')
print(table_classification)
```

```python
# Regression
results_regression = [
    ['Linear Regressor', -0.001, 14.103, 263.943],
    ['Decison Tree Regressor', -1.056, 18.963, 542.155],
    ['Random Forest Regressor', -0.032, 14.275, 272.079],
    ['Support Vector Regressor', -0.001, 14.106, 264.134]
]
```

```python
headers_regression=['Algorithms','r2_score','mean_absolute_error','mean_squared_error']
table_regression = tabulate(results_regression, headers=headers_regression, tablefmt='grid', floatfmt='.2f')
print(table_regression)
```

```python
# Eng yaxshi classification Algorithm
best_model_class = results_classification[0]
for model in results_classification:
    if model[4] >= best_model_class[4]:
        best_model_class = model
print(f"Eng yaxshi model Classifier uchun: ", best_model_class[0])
print(f"Eng yaxshi model metriging qiymati: ", best_model_class[4])
```

```python
# Eng yaxshi regression Algorithm
best_model_reg = results_regression[0]

for model_reg in results_regression:
    if model_reg[3] < best_model_reg[3]:
        best_model_reg = model_reg
print(f"Eng yaxshi model Regressor uchun: ", best_model_reg[0])
print(f"Eng yaxshi MSE qiymati: ", best_model_reg[3])
```
