# MEAN

```python
# 1.Mean orqali to'ldirish
# df['Survived'].fillna(df['Survived'].mean(),inplace=True)
# df['Survived'] = df['Survived'].fillna(df['Survived'].mean(),inplace=True)
df.fillna({'Survived': df['Survived'].mean()},inplace=True)
```

```python
df.isnull().sum()
```

# MODE

```python
# 2.Mode orqali to'ldirish
# df['Pclass'].fillna(df['Pclass'].mode()[0],inplace=True)
df['Pclass'] = df['Pclass'].fillna(df['Pclass'].mode()[0],inplace=True)
```

```python
df.isnull().sum()
```

# MEDIAN

```python
# 3.Median orqali to'ldirish
# df['Age'].fillna(df['Age'].median(),inplace=True)
df['Age'] = df['Age'].fillna(df['Age'].median(),inplace=True)
```

```python
df.isnull().sum()
```

# FIXED

```python
# 4.Fixed orqali to'ldirish
# df['Name'].fillna('Alli',inplace=True)
df['Name'] = df['Name'].fillna('Alli',inplace=True)
```

```python
df.isnull().sum()
```

# DROP

```python
# 5.Istalgan ustunni tashlab yuborish
df.drop('Sex',axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
# 6.Qatorlar bo'yicha to'ldirish
df.dropna(inplace=True)
```
