```python
# 2.Agar berilgan son musbat bo'lsa yaxshi, aks xolda yomon

a = int(input("Sonni kiriting"))

if a > 0:
    print('Yaxshi')
elif a == 0:
    print('Neytral son')
else:
    print('Yomon')
```

```python
# 4.Agar berilgan so'zning bosh harfi 'a' bo'lsa yaxshi, aks xolda yomon

gap = input("So`zni kiriting").lower() #--> lower() - bu so'z katta harfda bo'lgan holatlarda
# if len(gap) > 0 and gap[0] == 'a': #Agar hech narsa kiritilmasa xatolikni oldini olish uchun
if gap[0] == 'a':
    print('Yaxshi')
else:
    print('Yomon')
```

```python
# 5.Agar berilgan so'zning oxirgi xarfi `a` bo'lsa yaxshi, aks xolda yomon

soz = input("So'zni kiriting").lower() #--> lower() - bu so'z katta harfda bo'lgan holatlarda
# if len(soz) > 0 and soz[-1] == 'a': #Agar hech narsa kiritilmasa xatolikni oldini olish uchun
if soz[-1] == 'a':
    print('Yaxshi')
else:
    print('Yomon')
```

```python
# 6.Agar berilgan so'zning bosh xarfi `b` va oxirgi xarfi `a` bo'lsa yaxshi, aks xolda yomon

c = input("So'zni kiriting").lower() #--> lower() - bu so'z katta harfda bo'lgan holatlarda
# if len(c) > 0 and c[0] == 'b' and c[-1] == 'a': #Agar hech narsa kiritilmasa xatolikni oldini olish uchun
if c[0] == 'b' and c[-1] == 'a':
    print('Yaxshi')
else:
    print('Yomon')
```

```python
# 8.Berilgan sonni manfiy ekanligini isbotlang

f = int(input("Sonni kiriting"))

if f < 0:
    print('Son manfiy')
else:
    print('Son manfiy emas')
```

```python
# 9.Berilgan son nol(0) ekanligini isbotlang

o = int(input('Sonni kiriting'))

if o == 0:
    print('Nol(0) soni')
elif o > o:
    print('Musbat son')
else:
    print('Manfiy son')
```
