```python
import pandas as pd
df=pd.read_csv('business-financial-data-september-2025-quarter.csv')
```

# `For Loop` + `Missing` qiymatlarni to'ldirish

```python
# Missing qiymatlar turlari bo'yicha
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# 1-Mean
df['Period'] = df['Period'].fillna(df['Period'].mean())
```

```python
# 2-Mode
df['STATUS'] = df['STATUS'].fillna(df['STATUS'].mode()[0])
```

```python
# 3-Median
df['Data_value'] = df['Data_value'].fillna(df['Data_value'].median())
```

```python
# 4-Fixed
df['UNITS'] = df['UNITS'].fillna('6')
```

```python
# 5-Drop
df.drop('Suppressed',axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
# For Loop + Missing
```

```python
df.info()
```

```python
print(df.dtypes)
```

```python
# For Loop + Missing

for col in df.columns:
    if df[col].isnull().any():
        if df[col].dtype=='str':
            df[col] = df[col].fillna(df[col].mode()[0])
        else:
            df[col] = df[col].fillna(df[col].mean())
```

```python
# 1-usulda encoding qilish (Label=>harflar bo'yicha raqamlashtiradi)
df['Suppressed']=encoder.fit_transform(df['Suppressed'])
```

```python
# 2-usulda encoding qilish (One-Hot[0,1=>bo'yicha classlarni ustunga aylantiradi])
dummies=pd.get_dummies(df['STATUS'],prefix="STATUS",dtype=int)
```

```python
dummies
```

```python
df=pd.concat([df.drop(columns=['STATUS']),dummies],axis=1)
```

```python
df.info()
```

```python
# For Loop + Encoding

for col in df.columns:
    if df[col].dtype=='str':
        if df[col].nunique()<=6:
            dummies=pd.get_dummies(df[col],prefix=col,dtype=int)
            df=pd.concat([df.drop(columns=col),dummies],axis=1)
        else:
            df[col]=encoder.fit_transform(df[col])
```
