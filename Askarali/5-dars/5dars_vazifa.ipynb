```python
import pandas as pd
df=pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
df.tail()
```

# `MEAN` da o'rta arifmetik qiymat bilan to'ldirishdagi kod

```python
df.isnull().sum()
```

```python
df.info()
```

```python
# df['Survived'] = df['Survived'].fillna(df['Survived'].mean(),inplace=True) ----> bu usuli ham bor!
# df.mean({df['Survived']: 'Survived'}, inplace=True)
df.fillna({'Survived': df['Survived'].mean()}, inplace=True)  #bu varianti ham bor, iloji bo'lsa shundan foydalanishim pandas tominidan qo'llab quvvatlanadi.
```

```python
df.isnull().sum()
```

# `MODE` eng ko'p qaytarilgan qiymat (sonlar bir xil qaytarilsa mindan max ko'rinishida berib min qiymatni[0] qiymat deb olib beradi)

```python
# df['Pclass'] = df['Pclass'].fillna(df['Pclass'].mean(),inplace=True)
# df.fillna({"Pclass": df['Pclass'].mean()},inplace=True) ---> Bu qiymat 1,2,3 bo`lganligi uchun mode method bilan to`ldirish yaxshiroq
# df['Pclass'] = df['Pclass'].fillna(df['Pclass'].mode()[0],inplace=True)
df.fillna({'Pclass': df['Pclass'].mode()[0]},inplace=True)
```

```python
df.isnull().sum()
```

# `MEDIAN` eng o'rtadagi qiymat aniqlab berish uchun

```python
# df['Age'].fillna(df['Age'].median(),inplace=True)
# df['Age'] = df['Age'].fillna(df['Age'].median(),inplace=True)
df.fillna({'Age': df['Age'].median()},inplace=True) # ---> shu eng yaxshi foydalanish uchun
```

```python
df.isnull().sum()
```

# `FIXED` usulida qatorlarni to'ldirish

```python
# df['Name'].fillna('Ali',inplace=True)
# df['Name'] = df['Name'].fillna('Ali',inplace=True)
df['Name'] = df['Name'].fillna('Ali')
# df.fillna({'Name': 'Ali'},inplace=True)
```

```python
df.isnull().sum()
```

# `DROP` usuli orqali qatorlarni tashlab yuborish

```python
# qatorlar bo'yicha 1-holat
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
#drop ustun boyicha, malum bir ustunda juda kop(50% dan kop) qiymatlar tushib qolganda tashlab yuboramiz bu usutunni butunlay
df.drop('Name',axis=1,inplace=True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

# `ONE-HOT ENCODING`

```python
dummies=pd.get_dummies(df['Survived'],prefix='ustun',dtype=int)
```

```python
dummies
```

```python
df=pd.concat([df.drop(columns=['Survived']),dummies],axis=1)
```
