# 1- Berilgan son juft bo'lsa'even', toq bo'lsa 'odd' qaytaradigan function yozing

```python
def even_or_odd(number):
    condtion = 'even' if number % 2 == 0 else 'odd'
    return condtion 

even_or_odd(3)
```

# 3- List ichidagi nechta element borligini qataradigan function yozing 

```python
def count_elements(lst):
    return len(lst)

count_elements([1, 2, 5, 6, 7, 9, 10])
```

# 4- Sonni kvadratga oshiradigan funksiya yozing

```python
import math   # mathni import qlib olish orqali ham kvadratga oshirsa bo'ladi 

# buni number ** 2 orqali yoki number * number orqali ham bersa bo'ladi ammo men sal murakkabrog'ini tanladim
def square(number):
    return int(math.pow(number, 2))   # bu yerda agar int qilmasak bizga float qaytaradi masalan 25.0 degan

square(5)
```

# 5- List ichidagi eng katta sonni topadigan funksiya yozing

```python
# def find_max(lst):
#     return max(lst)

# find_max([3, 5, 6, 8, 10, 45])

# yoki bunday usuli ham bor ekan 

def find_max(lst):
    maximum = lst[0]  # bu yerda biz listning birinchi elementini vaqtingchalik eng kattasi deb oldik

    for num in lst:
        if num > maximum:   # yangi son eski maksimumdan kattami 
            maximum = num   # agar kattaroq bo'lsa maksimumni yangilaydi
    return maximum 

find_max([45, 67, 78, 55, 45, 33, 90])
```
