# Linear Regression based model 

```python
# import pandas to read the dataset 
import pandas as pd
```

```python
df = pd.read_csv('train.csv')
```

```python
df.info()
```

```python
df.isna().sum()
```

# Handling Missing Values

```python
def handle_miss_values(df):
    for col in df.columns:
        if df[col].isnull().any():
            if df[col].dtype == 'object':
                df[col].fillna(df[col].mode()[0], inplace=True)
            else:
                df[col].fillna(df[col].median(), inplace=True)
    return df
handle_miss_values(df)
```

```python
df.info()
```

# Encoding process

```python
df['Heart Disease'] = df['Heart Disease'].map({"Absence": 0, "Presence": 1})  # Only one columns is an objecty type so we've mapped it

```

```python
df.info()
```

```python
df.head()
```

```python
# Remove non-competetive elements
df.drop(['id', 'Sex'], axis=1, inplace=True)

```

```python
df.info()
```

# Scaling process

```python
from sklearn.preprocessing import MinMaxScaler
def do_scale(df):
    scaler = MinMaxScaler()
    num_cols = df.select_dtypes(include=['int64', 'float64']).columns.drop('Heart Disease')
    for col in df.columns:
        df[num_cols] = scaler.fit_transform(df[num_cols])
    return df
df = do_scale(df)
        
```

```python
df.head()
```

# Model Selection 

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop('Heart Disease', axis=1)
y = df['Heart Disease']
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size = 0.2, random_state = 42
)
```

# Linear Regression 

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr
```

```python
lr.fit(x_train, y_train)
```

# Prediction

```python
y_prediction = lr.predict(x_test)
```

```python
y_prediction[1:10]
```

# Evaluation 

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
r2 = r2_score(y_test, y_prediction)
mae = mean_absolute_error(y_test, y_prediction)
mse = mean_squared_error(y_test, y_prediction)
rmse = pow(mse, (1/2))

```

```python
print('r2: ', r2)
print('mae: ', mae)
print('mse: ', mse)
print('rmse: ', rmse)
```
