# 1-vazifa: Istalgan so'zni boshidagi va oxiridagi xarfini topadigan kod yozing

```python
s = input("Istalgan so'z kiriting: ")

birichi_harf = s[0]
oxirgi_harf = s[-1]

print("Birinchi harf: ", birichi_harf)
print("Oxirgi harf: ", oxirgi_harf)
```

# Berilgan ikki sonni musbat va juft ekanligini topadigan kod yozing

```python
a = int(input("Birinchi sonni kiriting: "))
b = int(input("Ikkinchi sonni kiriting: "))


if a > 0  and a%2 == 0: 
    print('Birinchi son musbat va juftdir')
else: 
    print("Birinchi son juft ham musbat ham emas!")
if b > 0  and b%2 == 0: 
    print('Ikkinchi son musbat va juftdir')
else: 
    print("Ikkinchi son juft ham musbat ham emas!")
```

# Berilgan so'zni oxiridan ikkinchi xarfini topadigan kod yozing

# Istalgan sonni toq yoki manfiy eknaligini aniqlaydigan misol yozing. 

```python
a = int(input("Birinchi sonni kiriting: "))


if a < 0  or a%2 == 1: 
    print('Bu son toq  yoki manfiydir')
else: 
    print("Bu son ham manfiy ham emas!") 
```

# Data Types va Operationlar misollarini ishlash

```python
# int
a = 5
b = 6
print(a+b)   # qo'shish
print(a-b)  # ayrish
print(a*b)   # ko'paytirish   
print(a/b)   # bo'lish 
print(a//b)  # butun qismini olish 
print(a%b)   # qoldigini olish 
print(a**b)  # darajasini olish 
```

```python
# float 
a = 3.6
b = 7.8
print(a+b)   # qo'shish
print(a-b)  # ayrish
print(a*b)   # ko'paytirish   
print(a/b)   # bo'lish 
print(a//b)  # butun qismini olish 
print(a%b)   # qoldigini olish 
print(a**b)  # darajasini olish 
```

```python
# str 
a = 'Hello'
b = ' World!'
print(a+b)
```

```python
# Boolean 
b = 19
print(b > 6)
```
