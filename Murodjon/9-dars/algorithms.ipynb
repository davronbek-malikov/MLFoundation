# Handling Missing Values jarayoni: Mode va M
























ean orqali

```python
# qolganlarini esa handling missing values qilib olamiz

for col in df.columns:
    if df[col].isnull().any(): 
        if df[col].dtype == 'str':
            df[col] = df[col].fillna(df[col].mode()[0], inplace=True)
        else: 
            df[col] = df[col].fillna(df[col].mean(), inplace=True)
```

```python
df.info()
```
