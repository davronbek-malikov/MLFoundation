# Uyga vazifa 2: Imtixondan o'tishni nazorat qilish uchun  condition tuzing

```python


math_exam_score = 70   
cmp_science_score = 60
overall_pass_threshold = 55

if math_exam_score >= overall_pass_threshold and cmp_science_score>=overall_pass_threshold:
    print("Natijalar: Har ikki fandan ham o'tdi")  # har ikkisidan o'tkan bo'lsa
elif math_exam_score>=overall_pass_threshold:
    print("Natija: Faqat Matematika imtihonidan o'tdi.") # faqat mathdan o'tkan bo'lsa
elif cmp_science_score>=overall_pass_threshold: 
    print("Natija: Faqat Kompyuter fanidagi imtohdan o'tdi.") # faqat computer sciencedan o'tkan bo'lsa
else: 
    print("Har ikki imtinondan ham o'ta olmadi.")  # har ikkisidan ham o'ta olmagan holdagi condition
```
