# Encoding 

```python
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
class Encoding_Scaling:
    def __init__(self, df):
        self.df = df
    def encodla(self):
        le = LabelEncoder()
        df = self.df

        for col in df.columns:
            if df[col].dtype == 'str':
                if df[col].nunique() <=5:
                    dummies = pd.get_dummies(df[col], prefix=col, dtype=int)
                    df = pd.concat([df.drop(columns=col), dummies], axis=1)
                else:
                    df[col] = le.fit_transform(df[col])

        self.df = df
        return df
    def scale_qil(self):
        scaler = MinMaxScaler()
        df = self.df
        num_cols = df.select_dtypes(include=['int64', 'float64']).columns
        for col in num_cols:
            df[col] = scaler.fit_transform(df[[col]])

        self.df = df
        return df

result = Encoding_Scaling(df)
df = result.encodla()
df = result.df
```

# Feature Engineering qilingan holat bilan qilib ko'ramiz.

```python
"""
Eng avvalo Feature Engineering bu yangi featurelar yaratish, ularni bir turdan boshqa turga transform qilish
va muhim featurelarni tanlash deganidir.
Bu model performancesini yaxshilaydi, va yangi featurelar model generalizationini oshiradi.
"""
```

```python
import pandas as pd
df = pd.read_csv('cars_dataset.csv')
```

```python
df.info()
```

```python
"""
bizda demak year, mileage va mpg bor --
year- bu chiqarilgan yili
mileage -- bu umumiy bosib o'tkan masofasi
mpg -- esa 1 gallonda bilan nechchi km yurishidir. --> 1 gallon ≈ 3.785 litr

"""
```

```python
df['moshina_yoshi'] = 2026 - df['year']  # bu moshinaning yoshi
df['bir_yildagi_masofa'] = df['mileage'] / df['moshina_yoshi'] # bu esa bir yilda nechchi km masofa bosib o'tkani
df['total_gallon_usage'] = df['mileage'] / df['mpg']  # bu esa umumiy masofani nechcha gallon bilan bosib o'tkanligi
```
