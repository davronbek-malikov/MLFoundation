```python
"""
Vazifa: Kitob degan klass yarating.

Atributlari: nomi, muallifi, yil va mavjud (bu boolean, ya'ni True yoki False bo'lsin).

Metodlari:

olish() — agar kitob kutubxonada bo'lsa, uni band qiladi (mavjud = False).

qaytarish() — kitobni kutubxonaga qaytaradi (mavjud = True).

malumot() — kitob haqidagi barcha ma'lumotlarni chiroyli ko'rinishda chiqaradi.

"""
```

```python
class Kitob:
    def __init__(self, nomi, muallifi, yil, mavjud=True):
        self.nomi = nomi
        self.muallifi = muallifi
        self.yil = yil
        self.mavjud = mavjud
    def olish(self, mavjud):
        if self.mavjud:
            self.mavjud = False
            print(f"{self.nomi}-kitobni oldim!")
        else:
            print(f"Kechirasiz, {self.nomi}-kitobi band ekan. Keyinroq olishingiz mumkin!")
    def qaytarish(self, mavjud):
        if not self.mavjud:
            self.mavjud = True
            print(f'{self.nomi} - bu kitobni qaytarilgan!')
        else:
            print(f"{self.nomi} - bu kitob allaqchon kutubxonada turibdi!")
    def malumot(self):
        holat = 'Kutubonada' if self.mavjud else 'Olingan'
        print(f"Kitob nomi: {self.nomi}, muallifi: {self.muallifi}, yili: {self.yil} va holati: {holat}")
kitob1 = Kitob("Dunyoning ishlari", "O'tkir Hoshimov", 1982)
kitob1.malumot()

```

```python
"""

Vazifa: Hamyon degan klass yarating.

Atributlari: egasi va balans (boshlang'ich qiymat 0 bo'lsin).

Metodlari:

kirim(miqdor) — balansga pul qo'shadi.

chiqim(miqdor) — agar balansda yetarli pul bo'lsa, pul ayiradi. Yetarli bo'lmasa, "Mablag' yetarli emas" deb xabar chiqaradi.

tarix() — qancha pul borligini ko'rsatadi

"""

```

```python
class Hamyon:
    def __init__(self, egasi, balans=0):
        self.egasi = egasi 
        self.balans = balans
    def deposit(self, miqdor):
        self.balans += miqdor
        print(f"{miqdor} won qo'shildi. Joriy balans: {self.balans} won")
    def withdrawal(self, miqdor):
        if miqdor <= self.balans:
            self.balans -= miqdor
            print(f'{miqdor} won yechildi. Qoldiq: {self.balans} won')
        else:
            print("Yetarli mablag' mavjud emas!")
    def history(self):
        print(f'Hamyon egasi: {self.egasi} Gafforov | Balans: {self.balans} won')
        
hamyon1 = Hamyon('Murodjon')
hamyon1.deposit(100000)
hamyon1.withdrawal(25000)
hamyon1.history()
```

```python
"""
Vazifa: Umumiy Xodim klassini va undan meros oluvchi Dasturchi klassini yarating.

Xodim atributlari: ism, maosh.

Dasturchi qo'shimcha atributi: til (masalan, Python, Java).

Vazifa: Dasturchi klassida maoshni 10% ga oshiruvchi bonus() metodini yozing.
"""
```

```python
# Inheritance --Vorislik 
class Xodim:
    def __init__(self, ism, maosh=0):
        self.ism = ism
        self.maosh = maosh
# Xodimdan (Ota classdan vorislik olamiz)
class Dasturchi(Xodim):
    def __init__(self, ism, maosh, til):
        # super() orqali ota classning __init__ methodini chiqamiz
        super().__init__(ism, maosh)
        self.til = til
        
        # Umumiy bonusni hisoblaymiz
        self.bonus_miqdori = self.maosh *0.1
        self.jami_maosh = self.maosh + self.bonus_miqdori 
    def malumot(self):
        print(f'Xodimning ismi: {self.ism} | Uning maoshi: {self.maosh} | Uning bilgan tillari {self.til} | Uning bonusi: {self.jami_maosh}')
dasturchi1 = Dasturchi('Asqarali', 10000000, 'Python, Java va Rust')
dasturchi1.malumot()
```

```python
"""

4. "Uy Hayvonlari" (Polimorfizm)

Vazifa: RobotHayvon degan klass yarating.

Bu klassda harakat() degan metod bo'lsin.

So'ngra UcharRobot va YurarRobot degan klasslarni RobotHayvondan meros qilib oling.

Har bir klassda harakat() metodini o'ziga xos tarzda (masalan, "Men uchayapman" yoki "Men yugurayapman") qayta yozing (Override qiling).

"""
```

```python
class RobotHayvon:
    def __init__(self, nomi):
        self.nomi = nomi
    def harakat(self):
        print('Uchar robotning nolasin tinglagin')

class UcharRobot(RobotHayvon):  
    # RobotHayvon ota classning ustidan yozamiz
    def harakat(self):
        print(f'{self.nomi}: Men uchayapman')
class YurarRobot(RobotHayvon):
    def harakat(self):
        print(f'{self.nomi}: Men yugurayapman')

robot1 = UcharRobot('Lochin')
robot2 = YurarRobot('Leopard')

robot1.harakat()
robot2.harakat()
```

```python
"""
5. "Do'kon Savatchasi" (Shopping Cart)

Vazifa: Savatcha degan klass yarating.

Atributi: mahsulotlar (bo'sh ro'yxat bo'lsin).

Metodlari:

qoshish(nomi, narxi) — mahsulotni lug'at ko'rinishida ({"nomi": "Olma", "narxi": 5000}) ro'yxatga qo'shadi.

umumiy_summa() — savatchadagi barcha mahsulotlar narxini hisoblab beradi.

tozalash() — savatchani bo'shatadi


"""
```

```python
class Savatcha:
    def __init__(self):
        self.mahsulotlar = []
    def qoshish(self, nomi, narxi):
        self.mahsulotlar.append({'nomi': nomi, 'narxi': narxi})
        print(f'{nomi} savatchaga qoshildi.')
    def umumiy_summa(self):
        jami = sum(m['narxi'] for m in self.mahsulotlar)
        print(f'Jami: {jami} sum')
    def tozalash(self):
        self.mahsulotlar = []
        print('Savatcha tozalandi')
    def malumot(self):
        print('Savatchadagi maxsulotlar: ')
        for m in self.mahsulotlar:
            print(f"-{m['nomi']}: {m['narxi']} sum")
savatcha1 = Savatcha()
savatcha1.qoshish('Olam', 12000)
savatcha1.qoshish('Nok', 10000)
savatcha1.qoshish('Banan', 19000)
savatcha1.umumiy_summa()
savatcha1.malumot()



    
```
