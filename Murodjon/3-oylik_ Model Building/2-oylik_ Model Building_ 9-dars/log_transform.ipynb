# Endi esa Log transformation va Feature Engineering 

```python
df.info()
```

```python
# Haftada o'rtacha qancha vaqt sarflaydi workout uchun
df['avg_workout'] = np.where(
    df['workouts_per_week'] == 0,  # agar workouts_per_week == 0 bo'lsa, 0 bo'lmasa normal hoatda hisoblanadi
    0,
    df['active_minutes'] / df['workouts_per_week']
)
# Haftalik umumiy harakat miqdori
df['total_activity'] = df['active_minutes'] * df['workouts_per_week']
# BMI va yosh kombinatsiyasi 
df['bmi_age'] = df['bmi'] * df['age']

```

```python
df.info()
```

```python
# Endi esa biz skewnessni tekshiramiz

num_cols = df.select_dtypes(include=['float32', 'int8'])
num_cols.skew()
```

```python
# heart_rate_resting bizda --manfiy
# continuous_exercise_days, efficiency_score, avg_workout , total_activity va bmi_age-- bizda 0.5 dan katta musbat
import numpy as np 
skewness = df.skew()
log_transform = skewness[(skewness >= 0.5)].index.tolist()
print(log_transform)
```

```python
# Logni qo'llaymiz 0.5 dan katta yoi teng bo'lgan columns uchun
for col in log_transform:
    df[col + '_log'] = np.log1p(df[col])
```

```python
df.info()
```

```python

# Manfiy heart_rate_resting bu ustnni squared transformation qilamiz
df['heart_rate_resting'] = df['heart_rate_resting'] **2

num_cols = df.select_dtypes(include=['float32', 'int8'])
num_cols.skew()
```

```python
# Class orqali DataPreprocess yasaymiz
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
class DataPreprocessing:
    def __init__(self, df):
        self.df = df
    
    def fill_miss_vals(self):
        for col in self.df.columns:
            if self.df[col].isnull().sum():
                if self.df[col].dtype == 'str':
                    self.df[col].fillna(self.df[col].mode()[0], inplace=True)
                else:
                    self.df[col].fillna(self.df[col].mean(), inplace=True)
        return self
    
    def encodla(self):
        le = LabelEncoder()
        for col in self.df.columns:
            if self.df[col].dtype == 'str':
                self.df[col] = le.fit_transform(self.df[col])
        return self
    
    def scale_qil(self, target_column='calorie_efficiency'):
        scaler = MinMaxScaler()
        # Raqamli ustunlarni aniqlaymiz
        num_cols = self.df.select_dtypes(include=['int64', 'float64']).columns.tolist()
        # raqamli ustunlarda target_column bo'lsa, uni olib tashlaymiz
        if target_column in num_cols:
            num_cols.remove(target_column)
    
        if num_cols:
            self.df[num_cols] = scaler.fit_transform(self.df[num_cols])
        
        return self


dp = DataPreprocessing(df)
dp.fill_miss_vals().encodla().scale_qil(target_column='calorie_efficiency').df
preprocessed_data = dp.df
```

```python
# preprocessed_data ni yana df'ga o'tkazib olish 
df = preprocessed_data
# x va y'ni berlgilab olish
x = df.drop('calorie_efficiency', axis=1)
y = df['calorie_efficiency']
# Model Selection 
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)
```
