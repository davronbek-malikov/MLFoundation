```python
# psoitive numlarni qiymatini negative qiymatlanri sonini topib ber

def sum_pos_count_neg(arr):
    pos_number = 0
    neg_number = 0
    if not arr:
        return []
    for i in arr:
        if i > 0:
            pos_number += i
        elif i < 0:
            neg_number += 1
    return list[pos_number, neg_number]
print(sum_pos_count_neg([9,8,8,-5,-6,-7,3,5]))
```

```python
# bmi yarastish, unda weight/height **2 ekan 
# underweight <=18.5
# Normal <= 25
# Overweight <=30 
# Obese > 30

def bmi(weight, height):
    bmi_calc = weight / height ** 2
    return (
    'Underweight' if bmi_calc <=18.5 else
    'Normal' if bmi_calc <=25 else
    'Overweight' if bmi_calc <=30 else
    'Obese'
    )
bmi_result = bmi(60, 1.78)
print(bmi_result)
```

```python
# Create a func with two arguments that will return an array of the first <n> multiplies <x>. 
# Example: x =1, n =10 ==> [1*1, 1*2, 1*3...1*10]
# or x=2, n=5 ==> [2*1, 2*2, ....2*5]

def count_by(x, n):
    ult_lst = []
    for i in range(1, n+1):   # bu yerda n+1 o'sha biz bergan songa teng bo'ladi chunki rangeda start kiradi lekin stop kirmaydi shunga +1 qosish orqali esa uni ham kirgizamiz
        ult_lst.append(x*i)  # bu esa x o'rniga masalan 5 kiritilsa uni har bir aylanada aylangan soniyga ko'paytirib beradi
    return ult_lst
rst_by = count_by(5, 9)
print(rst_by)
```

```python
# Given an array if integers your solution should find the smallest integer

def smallest(arr):
    find_smallest = arr[0]
    for char in arr:
        if char < find_smallest:
            find_smallest = char    # yoki shunchaki return min(arr) qilsak ham shuni qaytaradi
    return find_smallest
rst_smallest = smallest([5, 6, 9, 7, -7, 0])
print(rst_smallest)

```

```python
# Given a random non-negative number, you have to return the digits of this number within an array in a reverse order

def rev(n):
    num = []
    for i in str(n):
        num.append(int(i))
    num.reverse()
    return num 
print(rev(4234))
    


    # Yoki 
def rev_1(n):
    # Sonni stringga o'girib, teskari qilib, keyin har birini int qilib listga yig'amiz
    return [int(i) for i in str(n)][::-1]
print(rev_1(454543))
```

```python
# DNA ni RNA ga o'zgartirish, unda GCAT --> GCAU yoki 'T' bo'lsa 'U'ga o'zgartirihsh 

# 1-usuli 
def dna_to_rna(dna):
    result = ''
    if not dna:
        return ''
    for i in dna:
        if i == 'T':
            result += 'U'
        else:
            result += i
    return result 
rst1 = dna_to_rna('GTTS')
print(rst1)

# 2-usul

def dna_to_rna(dna):
    result = ''
    result = dna.replace('T', 'U')  # ancha clean code versiya
    return result 
rst2 = dna_to_rna('TTTT')
print(rst2)
```

```python
'''
Demak bid def needle() degan funksiya yozamiz
bu funkisya ichida keraskz har hil junk narsalar bor
- Lekin u array ichida bitta 'needle' bor 
Funksiya 'needle' ni topgandan keyin quyidagi ko'rinishda
string qaytaradi: 'found the needle at position X' degan 
Bu yerda X -> 'needle' joylashgan joyini ko'rsatadi

--Misol: ['hey', 'junk food', 'is here', 'needle is here']

'found the needle at position 3
'''

def find_needle(haystack):
    for index, element in enumerate(haystack):
        if element == 'needle':
            return 'found the needle at position ' + str(index)
result = find_needle(['hey', 'comeon', 'lets', 'eat', 'needle'])
print(result)
```

```python
"""
Write a function that removes the space from the string, then return the resulting string
For example: '9 asdsdsd asdsad ss sdsdd' --> '9asdsdsdasdsadsssdsdd'
"""
def no_space(x):
    result = ''
    for el in x:
        if el != ' ':
            result += el
    return result
rest = no_space('asds sdsad 00 sdsds 9')
print(rest)
```

```python
"""
Bizning futbol jamoaiz chempionatni tugatdi.
Jamoamizning har bir o'yin natijasi "x:y" ko'rinishidagi string(matn)
sifatida berilgan 
x - jamoamiz urgan gollar
y-raqib urgan gollar
Masalan --> ['3:2', '4:5', '3:0']
Agar x>y ---> 3 ochko ber
Agar x<y --> 0 ochko ber
Agar x=y teng bo'lsa 1 ochko ber.

"""

def gol_ur(games):
    total_points = 0

    for game in games:
        x, y = game.split(':')
        x = int(x)
        y = int(y)
        if x > y:
            total_points += 3
        elif x == y:
            total_points += 1
    return total_points
result = gol_ur(['4:2', '5:5', '2:0', '5:3', '0:9'])
print(result)
```

```python
"""
Simple, given a string of words, return the length of the shortest word(s)
String will never be empty and you don't need to account for different types


"""
def find_shorts(s):
    words = s.split()
    shortest = len(words[0])

    for word in words:
        if len(word) < shortest:
            shortest = len(word)
    return shortest
result = find_shorts('asds sad sdsdsd sd ssdsdd s')
print(result)
```

```python
"""
Musbat butun son 'n' berilgan 
shu sonnig nechta bo'luvchisi (divisor) borligini hisoblang

Masalan: n = 500 ga teng bo'lsa 500 % i == 0 bo'lsa += 1 ta qo'shib ketish kera
500--> 1, 5, 10, 50,.... bo'linadi buni qanday aniqlaymiz keling ko'ramiz
"""

def divisor(n):
    count = 0
    for i in range(1, n+1):
        if n%i == 0:
            count +=1
    return count
result = divisor(1000)
print(result)

```

```python
"""
Bob avtobus haydovchisi bo'lib ishlaydi. U juda mashxur bo'lib ketgan 
shuning uchun u juda ko'p yo'lovchilar uning avtobusiga chiqmoqchi bo'ladi
bazida avtobusda joy yetmay qoladi. 'Hamma yo'lovchilar sig'adimi yoki yo'q 
aniqlab beradigan funksiya qiling.
1 - cap --> bu avtobus sig'imi (haydovchisiz nechta)
2 - on --> hozir avtobus ichdia nechta odam bor
3 - wait --> bekatda kutib turgan odam soni

"""

def enough(cap, on, wait):
    empty = cap - on
    if wait <= empty:
        return 0  # agar joy bo'lsa 0 qaytaradi
    elif wait > empty: # agar 
        return wait - empty  
result = enough(100, 30, 90)
print(result)


# Or 
def enough(cap, on, wait):
    return max(0, wait-(cap-on))
res2= enough(200, 50, 170)
print(res2)
```

```python
"""
Complete the function that takes two integrers (a, b where a < b)
and return an array of all integers between the input parametres
including them. For example: a=1, b=4---> [1,2,3,4]

"""

def between(a,b):
    return list(range(a, b+1))
rest1 = between(4, 90)
print(rest1)

# OR 

def orta(a, b):
    return [result for result in range(a, b+1)]
rest2 = orta(5, 10)
print(rest2)
```

```python
"""
Write a function that checks if a given string (case sensitive)
is palindrome
For instance: madam -- madam the same backwards aas forwards
"""

def is_palindrome(s):
    origin = s.lower()
    reverse_origin = s[::-1].lower()
    if origin == reverse_origin:
        return True
    else:
        return False
res1 = is_palindrome('madam')
print(res1)
res2 = is_palindrome('socer')
print(res2)
```
