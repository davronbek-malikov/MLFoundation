# For Loopdan foydalangan holda va foydalanmagan holda missing qiymatlarni 5 ta usulda to'ldirish.

```python
import pandas as pd 
```

```python
df = pd.read_csv('employee_attrition_performance.csv')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.nunique()
```

```python
# For loopdan foydalanmgana holda 5 ta usulda tushirilgan qiymatlarni to'ldiramiz. 

# 1 - bu mean --> bu o'sha ustuni bo'yicha uning o'rta arifmetiki bilan to'ldirishdi faqat numerical qiymatlar uchun

df['environment_satisfaction'].fillna(df['environment_satisfaction'].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
# 2- bu mode --> bu eng ko'p takrorlangan qiymat bilan to'ldirishdi ham numerical ham catregorical qiymatlar uchun ishlatilaveradi.

df['attrition'].fillna(df['attrition'].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
# 3 - bu median --> bu o'rtaa turgan qiymat bilan to'ldirishdi faqat numerical qiymatlar uchun

df['relationship_satisfaction'].fillna(df['relationship_satisfaction'].median(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
# 4 - bu fixed --> bu qo'lda to'ldirish usuli, bunda oldin bilish yoki shu soxadan habardor bo'lish lozim

df['years_in_current_role'].fillna('Yes', inplace=True)
```

```python
df.isnull().sum()
```

```python
# 5-drop --bunda biz ikki hil usuldan foydalanamiz biri tushib qolgan qiymatlar oz bo'lganda 

df.dropna(inplace=True)  # Bunda dataframe ichidagi NaN (bo'sh) qiymatlari bor ustunlarni o'chirib tashlaydi.

```

```python
df.info()
```

```python
# 2-usuli esa aniq bir foizi yo'q lekin ustunimizda juda ko'p ma'lumotlar tanqisligi va suniylashib ketgan bo'lsa ishlatiladi. 
# Bunda ustun butunlay tashlanib yuboriladi. 
df.drop('performance_score', axis=1, inplace=True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Endi for loopdan foydalangan holda 5 ta usulda qiymatlari to'ldirib chiqamiz. 

for col in df.columns:
    if df[col].isnull().any():
        if df[col].dtype == 'object':
            df[col].fillna(df[col].mode()[0], inplace=True)
        else:
            df[col].fillna(df[col].mean(), inplace=True)

```

```python
df.isnull().sum()
```

```python
# Yoki median 

for col in df.columns:
    if df[col].isnull().any() and df[col].dtype != 'object':
        df[col].fillna(df[col].median(), inplace=True)
```

```python
df.isnull().sum()
```

```python
# Fixed holatida ko'ramiz

for col in df.columns:
    if df[col].isnull().any():
        if df[col].dtype == 'object':
            df[col].fillna('Unknown', inplace=True)  # matinli bo'lsa Unknown
        else: 
            df[col].fillna(-1, inplace=True) # sonli bo'lsa -1 bilan to'ldiradi.
```

```python
df.isnull().sum()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()   # bu barcha missing bo'lgan qatorlani o'chiradi. 
```

```python
df.isnull().sum()
```

```python
# agar faqat bitta ustun bo'yicha o'chirish kerak bo'lsa unda 
df = df.dropna(subset=['overtime'])  
```

```python
df.isnull().sum()
```

```python
# Agar 50% foizdan ortiq missing values bo'lsa unda drop columnsdan fodalanamiz bunda biz Threshold qo'yishimiz ishimizni osnonlashtiradi

threshold  = 0.5  # bu chegara qiymat 

for col in df.columns:
    if df[col].isnull().mean() < threshold:
        df.drop(columns=col, inplace=True)

```

```python
df.isnull().sum()
```

# For Loop + Encoding 

# bu yerda biz threshold ham qo'ysak maqsadga muvofiq sababi one-hot encoding nunique iloji boircha kichigini olishimiz kerak sababi dataset collapse bo'lmasligi uchun.

```python
df.nunique()
```

```python
from sklearn.preprocessing import LabelEncoder 

encoder = LabelEncoder()

encoder
```

```python
threshold = 5
for col in df.columns:
    if df[col].dtype == 'object':
        if df[col].nunique() <= threshold:
            dummies = pd.get_dummies(df[col], prefix=col, dtype=int)
            df = pd.concat([df.drop(columns=col), dummies], axis=1)
        else: 
            df[col] = encoder.fit_transform(df[col])
```
