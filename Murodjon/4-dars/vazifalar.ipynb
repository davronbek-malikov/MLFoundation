# Conditions 

```python
# Foydalanuvchidan kiritilgan sonni tekshirib, juft yoki toq ekanligini aniqlang.

juft_toq_son = int(input("Istalgan sonni kirting, uni juft yoki toqligini aniqlaymiz: "))

if juft_toq_son % 2 == 0: 
    result = "juft"
else: 
    result = "toq"

print(f"Kirtilgan son: {juft_toq_son}, {result}dir!")
```

```python
# Mahsulot narxi foydalanuvchidan kiritilsin.

# Agar narx 100 yoki undan katta boâ€˜lsa 10% chegirma berilsin, oxirgi narxni chiqaruvchi kod yozing.


maxsulot_narx = int(input('Maxsulot narxini kiriting: '))
foizli_chegirma = (maxsulot_narx * 10) / 100
final_narx = maxsulot_narx - foizli_chegirma


if maxsulot_narx > 100: 
    print(final_narx)
else: 
    print("Chegirmali harid uchun eng kamida 100 somdan ortiq bozorlik qiling") 
```
