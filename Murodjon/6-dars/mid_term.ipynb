```python
# 1. Agar berilgan son 2 ga bo’linsa yaxshi aks xolda yomon degan sozni ekrang chiqaring?

berilgan_son = int(input("Istalgan son kiriting: "))

if berilgan_son % 2 == 0: 
    print('Yaxshi')
else: 
    print('Yomon')
```

```python
# 2. Agar berilgan son musbat bo’lsa yaxshi aks xolda yomon degan sozni ekrang chiqaring?
berilgan_son = int(input("Istalgan son kiriting: "))

if berilgan_son > 0: 
    print('Yaxshi')
else: 
    print('Yomon')
```

```python
# 3. Agar berilgan 5 ga va 3 ga bolinsa yaxshi aks xolda yomon degan sozni ekrang chiqaring?
berilgan_son = int(input("Istalgan son kiriting: "))

if berilgan_son % 3 == 0 and berilgan_son % 5 == 0: 
    print('Yaxshi')
else: 
    print('Yomon')
```

```python
# 4. Agar berilgan so’zning bosh xarfi ‘a’ bolsa yaxshi aks xolda yomon degan sozni ekrang chiqaring 
berilgan_soz = input("Istalgan so'z kiriting: ")

if berilgan_soz[0] == 'a': 
    print('Yaxshi')
else: 
    print('Yomon')
```

```python
# 5. Agar berilgan so’zning oxirgi xarfi ‘a’ bolsa yaxshi aks xolda yomon degan sozni ekrang chiqaring 
berilgan_soz = input("Istalgan so'z kiriting: ")

if berilgan_soz[-1] == 'a': 
    print('Yaxshi')
else: 
    print('Yomon')
```

```python
# 6. Agar berilgan so’zning bosh xarfi ‘b’ va oxirgi xarfi ‘a’ bolsa yaxshi aks xolda yomon degan sozni ekrang chiqaring 
berilgan_soz = input("Istalgan so'z kiriting: ")

if berilgan_soz[0] == 'b' and berilgan_soz[-1] == 'a': 
    print('Yaxshi')
else: 
    print('Yomon')
```

```python
# 7. Berilgan sonning musbat ekanligini isbotlang
son = int(input('Berilgan son: '))

if son > 0: 
    print(f'{son} -musbat')
elif son == 0:
    print(f'{son} -na musbat va na manfiy; {son} neytral son')
else:
    print(f'{son} -musbat emas')
```

```python
# 8. Berilgan sonning manfiy ekanligini isbotlang
son = int(input('Berilgan son: '))

if son < 0: 
    print(f'{son} -manfiy')
elif son == 0:
    print(f'{son} -na musbat va na manfiy; {son} neytral son')
else:
    print(f'{son} -manfiy emas')
```

```python
# Berilgan sonning 0 ekanligini isbotlang 
son = int(input('Berilgan son: '))

if son == 0: 
    print(f'{son} ga teng')
else:
    print(f'{son}, 0 emas!')
```

# ML bo'yicha 

```python
import pandas as pd 
```

```python
df = pd.read_csv('employee_attrition_performance.csv')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Mean orqali to'ldiring 
df['relationship_satisfaction'].fillna(df['relationship_satisfaction'].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
# Mode orqali to'ldiring 
df['overtime'].fillna(df['overtime'].mode()[0], inplace=True)
```

```python
# Son orqali tushib qolgan qiymatlarni topish 
df.isnull().sum()
```

```python
df.info()
```

```python
# Median orqali to'ldiring 
df['environment_satisfaction'].fillna(df['environment_satisfaction'].median(), inplace=True)
```

```python
df.isnull().sum()
```

```python
# Fixed qiymat bilan to'ldiring --> Bu qo'lda to'ldirish usuli (oldin ko'rgan bo'lishi kerak yoki shu soxadan habari bo'lishi kerak)
df['attrition'].fillna('Yes', inplace=True)
```

```python
df.isnull().sum()
```

```python
# Istalgan ustunni tashlab yuboring 
df.drop('stock_option_level', axis=1, inplace=True)  # ustunda juda ko'p qiymatlar tushib qolganda sunilikdan qochish uchun ustunnni tashlab yuborish usuli
```

```python
df.isnull().sum()
```

```python
# Qatorlar bo'yicha to'ldiring 
df.dropna(inplace=True)
```
