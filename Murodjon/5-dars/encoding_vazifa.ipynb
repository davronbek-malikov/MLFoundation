# 5  ta usulda tushib qolgan qiymatlarni to'ldirish 

```python
import pandas as pd 
```

```python
df = pd.read_csv('Match_dataset.csv')
```

```python
df.info()
```

```python
df.nunique()
```

```python
df.isnull().sum()
```

```python
# Mean bu o'rta arifmetik qiymat numerical
 
df['Team_A_Ranking'].fillna(df['Team_A_Ranking'].mean(), inplace=True)    # Bu Team_A_Ranking ustunida 3 ta tushib qolgan qiymat bor edi 
```

```python
df.isnull().sum()
```

```python
# Mode eng ko'p takrorlangan qiymarlar bilan to'ldirish ham categorical ham numerical 

df['Toss_Winner'].fillna(df['Toss_Winner'].mode()[0], inplace=True)  # Toss_Winnerda esa 117ta missing values bor edi. 
```

```python
df.isnull().sum()
```

```python
# Median bu eng o'rtasidagi qiymat bilan to'ldirish --numberical  

df['_Ranking'].fillna(df['_Ranking'].median(), inplace=True)
```

```python
df.isnull().sum()
```

```python
# Fixed bu esa biz istagan qiymat bilan to'ldirishdir ham categorical ham numerical 

df['Match_ID'].fillna('2024-05-06', inplace=True)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
# Drop --bu qatorlarni o'chirib tashlash yoki tushurib yuborish 
df.dropna(inplace=True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# dropning ikkiinchi usuli butun bir ustunn tashlab yuborishda ishlatiladi bu asosan suniylikdan qochish uchun ishlatiladi.

df.drop('Venue', axis=1, inplace=True)
```

```python
df.isnull().sum()
```

# Shu dataset orqali One-hot Encoding va Label Encondinglarni ko'rib chiqamiz

```python
# One-hot encoding 
# birinchidan biz bu yerda df.nunique() ko'rib analiz qilishimiz kerak
df.nunique()
```

```python
# iloji boricha 5 dan kichigini ko'ramiz sababi kelgusida ustinlar soni qatorlar sonidan ortib collapsega duch kelmasligi uchun. 

dummies = pd.get_dummies(df['Toss_Decision'], prefix='col', dtype=int)

```

```python
dummies
```

```python
df = pd.concat([df.drop(columns=['Toss_Decision']), dummies], axis=1)  # orginal ustunni one-hot encoding qilingan holatda birlashtirish bo'ldi 
```

```python
df.info()
```

 # Label Encoding 

```python
# Bu asosan bizda label ya'ni tamg'a yoki belgilash ko'rinishida bo'ladi
# Categorical qiymatlarni alifo ketma-ketligida mos qiymarlar bilan belgilab chiqadi.

from sklearn.preprocessing import LabelEncoder   # bunda biz skearn kutubxonasidan preprocessing modulidan LabelEncoder classini chaqirib olamiz.
```

```python
encoder = LabelEncoder()  # bu yerda o'zgaruvchi berib LabelEncoderni chqariib olamiz
```

```python
encoder  
```

```python
df.info()
```

```python
df.nunique()
```

```python
# threshold o'laroq biz Stageni oldik sababi 5 dan kichikligi uchun 
df['Stage'] = encoder.fit_transform(df['Stage']) 
```
