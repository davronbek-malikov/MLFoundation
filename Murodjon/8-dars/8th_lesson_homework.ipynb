# Homework for Lesson 8: using a for loop for encoding and scaling

# For Loop orqali tushib qolgan qiymatlarni to'ldiring 5 ta usuldan foydalangan holda

```python
import pandas as pd
```

```python
df = pd.read_csv('price_of_healthy_diet_clean.csv')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Mean va Mode ni for loop orqali avtomatlashtiramiz

for col in df.columns:
    if df[col].isnull().any():
        if df[col].dtype == 'object':
            df[col].fillna(df[col].mode()[0], inplace=True)
        else:
            df[col].fillna(df[col].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df.isnull().sum()
```

```python
# Yoki median 

for col in df.columns:
    if df[col].isnull().any() and df[col].dtype != 'object':
        df[col] = df[col].fillna(df[col].median(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
# Fixed bilan ishlash 
for col in df.columns: 
    if df[col].isnull().any():
        if df[col].dtype == 'object':
            df[col].fillna('Tashlab yuborildi', inplace=True)  # bunda object bo'lsa qo'lda Tashlab yuborildi deb yozib chiqadi avtoatik tarzda
        else: 
            df[col].fillna(-1, inplace=True) # bunda esa numerical valuelarni avtomatik ravishda -1 bilan to'ldiradi
```

```python
df.isnull().sum()
```

```python
# Droping qator bo'ylab tashlab yuborishi for loopsiz qulayroq, tezroq va tozaroq
df = df.dropna()   # bu barcha missing bo'lgan qatorlani o'chiradi. 
```

```python
df.isnull().sum()
```

```python
df.isnull().sum()
```

```python
# agar faqat bitta ustun bo'yicha o'chirish kerak bo'lsa unda 
df = df.dropna(subset=['country_code'])  
```

```python
df.isnull().sum()
```

```python
# Agar 50% foizdan ortiq missing values bo'lsa unda drop columnsdan fodalanamiz bunda biz Threshold qo'yishimiz ishimizni osnonlashtiradi

threshold  = 0.5  # bu chegara qiymat 

for col in df.columns:
    if df[col].isnull().mean() > threshold:    # bu yerda mean() ishlayotkani sababi pandasda bu True-1 False-0 qaytaradi va umumiy ulushdagi mean() 0.5 dan baland bo'lsa drop qiladi.
        df.drop(columns=col, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.info()
```

# For Loop + Encoding 

```python
df.nunique()
```

```python
# bu yerda biz kichikroqg'ini olishligimiz maqsadga muvofiq sababi datasetimiz collapse bo'lmasligi uchun. bunda bizga Treshodl yodam beradi. 
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
encoder
```

```python
threshold = 5   #chegara 
  
for col in df.columns:
    if df[col].dtype == 'object':
        if df[col].nunique() <=threshold:
            dummies = pd.get_dummies(df[col], prefix=col, dtype=int)
            df = pd.concat([df.drop(columns=col), dummies], axis=1)
        else: 
            df[col] = encoder.fit_transform(df[col])
        
```

```python
df.isnull().sum()
```

```python
df.info()
```

# For Loop + Scaling 

```python
# birinchi 3 ta scaling classlari import qilib olamiz 
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler

```

```python
scaler = MinMaxScaler()  # bunda asosan [0,1] oraliqda balansladi
scaler 
```

```python
for col in df.columns:
    if df[col].dtype != 'object':
        df[col] = scaler.fit_transform(df[[col]])  # shu yerda iktalik qovus ishlatish kerak
```

```python
df.head()
```

```python
# StandardScaler 
scaler = StandardScaler()
scaler
```

```python
for col in df.columns:
    if df[col].dtype != 'object':
        df[col] = scaler.fit_transform(df[[col]])    # bunda faqat manfiy tarafga qiymatlarni balanslaydi
```

```python
# RobustScaler
scaler = RobustScaler()
scaler
```
