```python
# Berilgan ikki sonni musbat va juft ekanligini topadigan kod yozing

def musbat_juftmi(x):
    return x > 0 and x % 2 == 0

a = int(input("Birinchi raqamni kiriting: "))
b = int(input("Ikkinchi raqamni kiriting: "))

print("Raqamingiz musbat va juft" if musbat_juftmi(a) else "Raqam musbat emas va juft emas")
print("Raqamingiz musbat va juft" if musbat_juftmi(b) else "Raqam musbat emas va juft emas")
```

```python
# Istalgan sonni toq  yoki manfiy ekanligini tekshiradigan kod yozing

def odd_negative(x):
    if x > 0 and x % 2 == 0:
        print("raqam musbat va juft")
    elif x < 0 and x % 2 != 0:
        print("raqam manfiy va toq")
    elif x > 0 and x % 2 != 0:
        print("raqam musbat va toq")
    else: 
        print("boshqa raqam kiriting.")

n = int(input("Raqam kiriting: "))
odd_negative(n)
```
