```python
from preprocessing import Preprocessing
import pandas as pd
import os
```

```python
df = pd.read_csv("../data/raw/Teen_Mental_Health_Dataset.csv")
```

```python
df.info()
```

```python
df.isnull().sum()
```

## **Data preprocessing using modular programming**

```python
from preprocessing import Preprocessing

cleaned_data = Preprocessing(df)
cleaned_data.handling_missing_values().encoding().scaling()
df_cleaned = cleaned_data.df
```

```python
df_cleaned.isnull().sum()
```

```python
df_cleaned.info()
```

## **Saving preprocessed data into new folder**

```python
output_folder = "../data/preprocessed"
os.makedirs(output_folder, exist_ok=True)
output_path = os.path.join(output_folder, "preprocessed_df.csv")
df_cleaned.to_csv(output_path, index=False)
```

```python
df_cleaned.sample(10)
```

## **Random Forest Regressor**

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np


x = df_cleaned.drop("academic_performance", axis=1)
y = df_cleaned["academic_performance"]

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)

rf = RandomForestRegressor()

rf.fit(x_train, y_train)

y_pred = rf.predict(x_test)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(mae)
print(mse)
print(rmse)
print(r2)

```

##  **Random Forest Regressor without scaling**

```python
df = pd.read_csv("../data/raw/Teen_Mental_Health_Dataset.csv")
```

```python
df.info()
```

```python
from preprocessing import Preprocessing

cleaned_data = Preprocessing(df)
cleaned_data.handling_missing_values().encoding()
df = cleaned_data.df
```

```python
df.sample(10)
```

```python

x = df.drop("academic_performance", axis=1)
y = df["academic_performance"]

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)

rf = RandomForestRegressor()

rf.fit(x_train, y_train)

y_pred = rf.predict(x_test)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(mae)
print(mse)
print(rmse)
print(r2)

```

## **K-fold cross validation**

```python
from sklearn.model_selection import KFold, cross_val_score

kf = KFold(n_splits=5, shuffle=True, random_state=42)

scores = cross_val_score(rf, x,y,cv=kf, scoring="r2")
print(scores.mean())

from sklearn.metrics import make_scorer

mae=make_scorer(mean_absolute_error,greater_is_better=False)
scores=cross_val_score(rf,x,y,cv=kf,scoring=mae)
print(-scores)
```
