## **ENSEMBLE METHODS**

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
```

```python
df = pd.read_csv("../data/ev_market_2026.csv")
```

```python
df.info()
```

## **Data Preprocessing**

```python
# Handling missing values

df.isnull().sum()
```

```python
# Encoding

def encoding(df):
    for col in df.columns:
        encoder = LabelEncoder()
        if df[col].dtype == "object":
            df[col] = encoder.fit_transform(df[col])
    return df

df = encoding(df)
df
```

```python
df.info()
```

```python
# Scaling 

def scaling(df):
    scaler = MinMaxScaler()
    num_cols = df.select_dtypes(include=["float64", "int32", "int64"]).columns.drop("price_usd")
    df[num_cols] = scaler.fit_transform(df[num_cols])
    return df

df = scaling(df)
df
```

## **Random Forest Regressor**

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

x = df.drop("price_usd", axis=1)
y = df["price_usd"]

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)

model = RandomForestRegressor()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)


mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(mae)
print(mse)
print(rmse)
print(r2)
```

```python
# K-fold cross validation
from sklearn.model_selection import KFold, cross_val_score

kf = KFold(n_splits=5, shuffle=True, random_state=42)

scores = cross_val_score(model, x,y,cv=kf, scoring="r2")
print(scores.mean())

from sklearn.metrics import make_scorer

mae=make_scorer(mean_absolute_error,greater_is_better=False)
scores=cross_val_score(model,x,y,cv=kf,scoring=mae)
print(-scores)
```

## **AdaBoostRegressor**

```python
from sklearn.ensemble import AdaBoostRegressor

ada = AdaBoostRegressor(
    n_estimators=100,
    learning_rate=0.1,
    random_state=42
)

ada.fit(x_train, y_train)
y_pred = ada.predict(x_test)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(mae)
print(mse)
print(rmse)
print(r2)
```

```python
# K-fold cross validation
from sklearn.model_selection import KFold, cross_val_score

kf = KFold(n_splits=5, shuffle=True, random_state=42)

scores = cross_val_score(ada, x,y,cv=kf, scoring="r2")
print(scores.mean())

from sklearn.metrics import make_scorer

mae=make_scorer(mean_absolute_error,greater_is_better=False)
scores=cross_val_score(ada,x,y,cv=kf,scoring=mae)
print(-scores)
```

## **Gradient Boosting**

```python
from sklearn.ensemble import GradientBoostingRegressor

gb = GradientBoostingRegressor(
    n_estimators=100,
    learning_rate=0.1,    
    max_depth=3,
    subsample=0.8,        
    random_state=42
)

gb.fit(x_train, y_train)
y_pred = gb.predict(x_test)

ada.fit(x_train, y_train)
y_pred = ada.predict(x_test)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(mae)
print(mse)
print(rmse)
print(r2)
```

```python
# K-fold cross validation
from sklearn.model_selection import KFold, cross_val_score

kf = KFold(n_splits=5, shuffle=True, random_state=42)

scores = cross_val_score(gb, x,y,cv=kf, scoring="r2")
print(scores.mean())

from sklearn.metrics import make_scorer

mae=make_scorer(mean_absolute_error,greater_is_better=False)
scores=cross_val_score(gb,x,y,cv=kf,scoring=mae)
print(-scores)
```

## **XGboost**

```python
import xgboost as xgb

xgb_model = xgb.XGBRegressor(
    n_estimators=100,
    learning_rate=0.1,
    max_depth=6,
    subsample=0.8,
    colsample_bytree=0.8,  # features per tree
    reg_alpha=0.1,         # L1 regularization
    reg_lambda=1.0,        # L2 regularization
    random_state=42
)

xgb_model.fit(x_train, y_train)

y_pred = xgb_model.predict(x_test)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(mae)
print(mse)
print(rmse)
print(r2)
```

```python
# K-fold cross validation
from sklearn.model_selection import KFold, cross_val_score

kf = KFold(n_splits=5, shuffle=True, random_state=42)

scores = cross_val_score(xgb_model, x,y,cv=kf, scoring="r2")
print(scores.mean())

from sklearn.metrics import make_scorer

mae=make_scorer(mean_absolute_error,greater_is_better=False)
scores=cross_val_score(xgb_model,x,y,cv=kf,scoring=mae)
print(-scores)
```
