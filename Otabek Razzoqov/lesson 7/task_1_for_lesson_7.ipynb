# WORKING WITH DATASET

```python
import pandas as pd 
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
encoder
```

```python
df = pd.read_csv("IMDB_Movies_Dataset.csv")
df.isnull().sum()
```

```python
df.head(5)
```

```python
df.info()
```

#### mean() , median(), mode(), fixed(), drop()

```python
df['Average Rating'].fillna(df['Average Rating'].mean(), inplace=True)
df['Metascore'].fillna(df['Metascore'].median(), inplace = True)
df['Writer'].fillna(df['Writer'].mode()[0], inplace = True)
df.drop(columns = ['Unnamed: 0'], axis = 1, inplace = True)
df.drop(columns = ['Average Rating.1'], axis = 1, inplace = True)
df.drop(columns = ['Writer.1'], axis = 1, inplace  = True)

```

```python
df.sample(1)
```

#### Handling missing data with for loop

```python
df.columns
```

```python
for col in df.columns:
    if df[col].isnull().any():
        if df[col].dtype != 'object':
            if df[col].nunique() <= 5:
                df[col].fillna(df[col].mean(), inplace = True)
            elif df[col].nunique() > 5:
                df[col].fillna(df[col].meadian(), inplace = True)
        elif df[col].dtype == "object":
            df[col].fillna(df[col].mode()[0], inplace = True)

```

```python
df.info()
```

### Encoding part
#### One-hot encoding and Label encoding

```python
for col in df.columns:
    if df[col].dtype == "object":
        if df[col].nunique() <= 5:
            dummies = pd.get_dummies(df[col], prefix=col, dtype=int)
            df = pd.concat([df.drop(columns=col), dummies], axis = 1)
        else:
            df[col] = encoder.fit_transform(df[col])
```
