### Imtihondan o'tishni nazorat qilish uchun  condition tuzing.


```python
baho = int(input("Bahongizni kiriting:"))
if baho > 86:
    print("A'lo")
elif baho > 71 and baho < 86:
    print("Yaxshi")
elif baho > 56 and baho < 71:
    print("Qoniqarli")
else:
    print("Qayta topshirish yoki bahoni xato kiritdingiz!")
```

### Passport olish uchun condition tuzing.

```python
Age = int(input("Enter your age:"))
if Age > 16:
    print("You can now have a passport!")
elif Age < 16 and Age >0:
    print(f"You have to wait for {16 - Age} years to get your passport!")
else:
    print("You entered your age incorrectly!")
```
