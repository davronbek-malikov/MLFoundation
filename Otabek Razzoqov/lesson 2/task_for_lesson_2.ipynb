## Writing a code that finds the first and last letter of a given word.

```python
word = str(input("write any word:"))
print(f"The first letter of a given word is: {word[0]}")
print(f"The last letter of a given word is: {word[-1]}")
```

## Writing a code that finds if a given number is odd or even number.

```python
number = int(input("Choose a number:"))

if number > 0 and number%2 == 0:
    print(f"{number} is even and positive")
elif number < 0 and number%2 ==0:
    print(f"{number} is even and negative")
elif number > 0 and number%2 != 0:
    print(f"{number} is odd and positive")
else:
    print(f"{number} is odd and negative")
```

## Writing a code that returns the second-to-last character of a given word

```python
word = str(input("write any word:"))
if len(word)>1:
    print(f"The second-to-last letter of a given word is: {word[-2]}")
else:
    print(f"The word has only one letter and it is '{word}'")

```
