# ML CODING

```python
import pandas as pd
```

```python
df = pd.read_csv('cleaned_data.csv')
df.isnull().sum()
```

```python
df.info()
```

```python
df.sample(5)
```

# Filling with mode()

```python
df['os'].fillna(df['os'].mode()[0], inplace=True)
df['os'].isnull().sum()

df['core_type'].fillna(df['core_type'].mode()[0], inplace=True)
df['core_type'].isnull().sum()

df['resolution'].fillna(df['resolution'].mode()[0], inplace=True)
df['display_type'].fillna(df['display_type'].mode()[0], inplace=True)
df['memory_card_type'].fillna(df['memory_card_type'].mode()[0], inplace=True)
```

# Filling with mean()

```python
df['front_camera_mp'].fillna(df['front_camera_mp'].mean(), inplace=True)
df['front_camera_mp'].isnull().sum()

df['clock_ghz'].fillna(df['clock_ghz'].mean(), inplace=True)
df['refresh_rate_hz'].fillna(df['refresh_rate_hz'].mean(), inplace=True)
df['rear_camera_max_mp'].fillna(df['rear_camera_max_mp'].mean(), inplace=True)
```

# Filling with median()

```python
df['memory_card_max_gb'].fillna(df['memory_card_max_gb'].median(), inplace=True)
df['memory_card_max_gb'].isnull().sum()

df['battery_mah'].fillna(df['battery_mah'].median(), inplace=True)
df['fast_charge_w'].fillna(df['fast_charge_w'].median(), inplace=True)
df['screen_size_in'].fillna(df['screen_size_in'].median(), inplace=True)
```

# Filling with fixed() value

```python
df['rear_camera_count'].isnull().any()
```

```python
df['rear_camera_count'].fillna(1, inplace=True)
df['rear_camera_count'].isnull().any()
```

# Dropping a column 

```python
df.drop('memory_card_max_gb', axis=1, inplace=True)
```
