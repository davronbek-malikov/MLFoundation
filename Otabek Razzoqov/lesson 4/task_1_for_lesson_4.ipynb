# mean()
### Replacing missing values with the average of the column.

### ✔ When to use:
###  Numerical data
###  Data is normally distributed
###  No strong outliers
###  ⚠ Problem:
###  Mean is sensitive to outliers.

```python
df['Inflation (CPI %)'].fillna(df['Inflation (CPI %)'].mean(), inplace=True)
```

```python
df.info()
```

# median()
### ✔ What it does:     Replaces missing values with the middle value.
### df['salary'].fillna(df['salary'].median(), inplace=True)
### ✔ When to use:
### Numerical data
### Data has outliers
### Skewed distribution
### ✅ More robust than mean

```python
df['GDP (Current USD)'].fillna(df["GDP (Current USD)"].median(), inplace=True)
df['GDP (Current USD)'].value_counts().sum()
```

#   mode()
### ✔ What it does:       Replaces missing values with the most frequent value.
###  df['gender'].fillna(df['gender'].mode()[0], inplace=True)
### ✔ When to use:
###  Categorical data
### Discrete variables

```python
df['country_name'].fillna(df['country_name'].mode()[0], inplace=True)
df['country_name'].value_counts().sum()
```

```python
df.info()
```

# Fixed Value Replacement
### ✔ What it does:   Replace missing values with a constant value.

### df['score'].fillna(0, inplace=True)
### ✔ When to use:

### When missing means something specific
### Business logic requires fixed value
### Feature engineering trick (e.g., -1 for missing)

```python
df.head(10)
```

```python
df.isnull().sum()
```

```python
data = df['Public Debt (% of GDP)'].mean()
df['Public Debt (% of GDP)'].fillna(data, inplace=True)
```

```python
df.isnull().sum()
```
