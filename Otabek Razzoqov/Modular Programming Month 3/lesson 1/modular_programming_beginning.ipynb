## **Class Examples**

```python
class BankAccount:
    def __init__(self, current_balance):
        self.current_balance = current_balance
    
    def deposit(self, amount):
        if amount <= 0:
            return f"Deposit amount must be positive"
        self.current_balance +=amount
        return self.current_balance
    
    def withdraw(self, amount):
        if amount <= 0:
            return f"Withraw amount must be positive"
        if amount > self.current_balance:
            return f"Insufficient funds"
        balance_before = self.current_balance
        self.current_balance -= amount
        return balance_before
    
    def get_balance(self):
        return self.current_balance
    

    
```

```python
mening_pulim = BankAccount(10000)

print(mening_pulim.deposit(5000))
```

```python
print(mening_pulim.withdraw(7000))
print(mening_pulim.get_balance())
```

```python
class StudentGrades:
    def __init__(self, name, grades):
        self.name = name
        self.grades = grades

    def get_lowest(self):
        return min(self.grades)
    
    def get_gighest(self):
        return max(self.grades)
    
    def get_average(self):
        return sum(self.grades)/len(self.grades)
```

```python
men = StudentGrades("Otabek", [87,76,98,65,89])

men.get_average()
men.get_gighest()
men.get_lowest()
```
